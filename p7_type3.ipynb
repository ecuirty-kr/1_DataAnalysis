```python
##### 3유형 6 까지 하고 1 유형 기출 조금만 풀고 마무리하기.
#### 문제 조건 : 로지스틱 회귀 - 조개 정보를 나타낸 데이터이다. 주어진 데이터에서 300개 중 앞에서부터 210개는 train으로, 90는 test 데이터로 나눈다.
#### 모델을 학습할 때는 train 데이터를 사용하고, 예측할 때는 test 데이터를 사용한다. 모델은 다중 로지스틱 회귀를 써서 성별(gender)을 예측하되,
#### 패널티는 부과하지 않는다
#### 종속변수 : gender(0:암컷 1:수컷)    / data : clam.csv

## 문제 1. weight를 독립변수, gender를 종속변수로 사용해 로지스틱 회귀 모델을 만들고, weight 변수가 한 단위 증가할 때 수컷일 오즈비 값? (소수 넷째 반올림)
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/clam.csv")
# 데이터 확인
print(df.head(4))

# 문제 요구사항 : 210 기준으로 데이터 분할
train = df[:210].copy()
test = df[210:].copy()
print(train.shape, test.shape) # 분할 데이터 크기 맞는지 확인 (210, 6) (90, 6)

# 2. 로지 회귀 모델 적합
from statsmodels.formula.api import logit
model = logit('gender~weight', data=train).fit()
# print(model.summary()) ## 통계 요약 확인
# weight 한 단위 gender 오즈비 / 오즈비!!! numpy!!!!!!!!!!!!!!!!
import numpy as np
odds_ratio = np.exp(model.params['weight']) # 한 단위니까 곱 (X)
print(round(odds_ratio, 4)) # 결과 : 1.0047
# 답안 : 1.0047

## 문제 2. gender를 종속변수로 하고 나머지 변수들(age, length, diameter, height, weight)을 독립변수로 사용하는 로지스틱 회귀 모델을
## 적합시킨 후 잔차 이탈도(Residual Deviance)를 계산하시오 (반올림 소수 둘째)
# 새 모델 적합
new_model = logit('gender~age+length+diameter+height+weight', data=train).fit()
print(new_model.summary()) # 통계 확인

# 잔차 이탈도 (?) - 잔차부터 구해보자. 잔차 = 실제 - 예측
# pred = model.predict(test) # 구하는 법을 모르겠다
# 잔차 이탈도 : 로그 우도 값에 -2 를 곱한 값 - 모델의 적합도를 평가하는 지표
out = new_model.llf * -2  ## 로그 우도=model.llf : 모델이 주어진 데이터에 얼마나 잘 맞는지를 나타냄
print(round(out, 2)) # 결과 : 286.93
# 답안 : 286.93

## 문제 3. 독립변수 weight 만 사용해 학습한 모델에서 test 데이터의 gender 를 예측하고, Error Rate(오류율) 구하기 (소수 셋째 반올림)
# 아까 문제 1에서 만든 모델이름과 2에서 만든 모델 이름 다르게 설정 (여기서 1 모델 사용)
# 예측 predict() : 예측은 test 데이터 사용
# pred = model.predict(test)
# print(pred) # 예측값 확인    **************** 3 유형에서 보지 못한 문제 유형 **********************

# 모범 답안 : 오류율 = (1-정확도) [정확도 라이브러리 accuracy_score]
from sklearn.metrics import accuracy_score
target = test.pop('gender')  ## 예측에 사용할 데이터이므로 test에서 종속변수('gender')꺼냄
pred = model.predict(test)   ## test 데이터로 예측 수행
# pred 값의 범위는 0~1으로, 0.5 보다 큰 값이어야만 관측치가 1에 속함 -> 0.5 보다 큰 pred 값만 필터링
pred = (pred > 0.5).astype(int) # 정수형으로

# 오류율 구하기 : 정확도 구하기 - (1-정확도) = 오류율
accuracy = accuracy_score(target, pred)  ## 평가지표 계산에 pred 를 넣는게 아니라 target, pred를 넣어서 'gender'를 예측(pred) 하는것임.!
error = (1-accuracy)
print(round(error, 3)) # 결과 : 0.478
# 답안 : 0.478

## 그래도. 이정도면 감점은 되더라도 완전 오답은 안 나오겠지
## 6 회차 풀고. 1 유형 조금만. 하면 몇 시에 끝나지    ...  아 3 유형 문제 두 개지
```

```python
### 문제 조건: 선형 회귀 / data: system_cpu.csv
## 문제 1. ERP와 가장 상관 관계가 높은 값을 구하시오 (소수 셋째 반올림)
# 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/system_cpu.csv")
print(df.head(5)) # 데이터 확인

# # 선형 회귀 : 상관 관계 - 상관계수 확인
# from statsmodels.formula.api import ols
# model = ols('ERP~Feature1+Feature2+Feature3+CPU', data=df).fit()
# print(model.corr()) ## ???? 이게 아닌가.

# ERP 와 각 변수 사이의 상관계수 계산
corr_matrix = df.corr() ## corr() 함수는 데이터 프레임(df)에서 함수 실행 df.corr() -> 상관계수를 행렬 형태로 조회
# ERP 와 다른 변수들과의 상관계수 출력
erp_corr = corr_matrix['ERP'].sort_values(ascending=False) # 상관계수 관계를 내림차순으로 정렬
print(erp_corr) # 가장 상관 관계가 높은 값 : Feature3   0.882194 -> 소수 셋째 반올림 : 0.882
# 답안 : 0.882

## 문제 2. CPU 컬럼이 100 미만인 것만 찾아 ERP를 종속변수, 나머지 변수들을 독립변수로 설정해 선형 회귀 모델을 만들고
## 적합한 결정 계수를 구하시오 (소수 셋째 반올림)
# 선형 회귀 ols
from statsmodels.formula.api import ols
# ***************************** CPU 컬럼 100 미만 데이터 필터링 과정 안 함 **************************
# filtered_df = (df['CPU'] < 100).astype(int)  // 틀림
filtered_df = df[df['CPU'] < 100] ## df에서 df['CPU'] 값이 100 미만인 것들만 필터링
model = ols('ERP~Feature1+Feature2+Feature3+CPU', data=filtered_df).fit()
# ***************************************************************************************************
print(model.summary())
print(round(model.rsquared, 3)) # 결정계수 rsquared / 소수 셋째 반올림 : 0.755
# 답안 : 0.755    ** summary() 에서 R-squared 값으로도 확인 가능

## 문제 3. 위에서 만든 모델에서 독립 변수 중 p-value가 가장 높은 값을 구하시오 (소수 셋째 반올림)
# 가장 높은 값 idxmax / 가장 높은 인덱스 전체 정보에서 p-value 찾아도 가능
result = model.pvalues.idxmax()
print(result) # 결과 : Feature1
print(round(model.pvalues['Feature1'],3)) # 결과 : 0.684
# 답안 : 0.822   // 틀림
# 모범 답안 : summary() 결과에서 p-value 가장 높은 값 직접 확인
# 답안 : 0.684

## 적용할 train 데이터 잘못(필터링 생략) 설정하면 모든 값이 틀어지므로 유의.!!! 문제 잘 읽기.
```

### 문제1-1

```python
import pandas as pd
# df = pd.read_csv("clam.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/clam.csv")

# 데이터셋 분할
print(df.shape)
train = df.iloc[:210]
test = df.iloc[210:]
print(train.shape)

print(train.head())
```

```python
from statsmodels.formula.api import logit
import numpy as np

# 1) 로지스틱 회귀 모델 생성 및 학습
model = logit("gender ~ weight", data=train).fit()

# 2) 오즈비 계산
odds_ratio = np.exp(model.params['weight'])
print(round(odds_ratio, 4))
```

### 문제1-2

```python
# 1) 로지스틱 회귀 모델 생성 및 학습
formula = "gender ~ age + length + diameter + height + weight"
model = logit(formula, data=train).fit()

# 2) 잔차이탈도
print(round(-2 * model.llf,2))
```

### 문제1-3

```python
from sklearn.metrics import accuracy_score

# 1) test데이터를 사용해 예측 (0.5 미만: 0, 0.5 이상 1)
model = logit("gender ~ weight", data=train).fit()
target = test.pop('gender')
pred = model.predict(test)
pred = (pred > 0.5).astype(int)

# 2) 실제 값과 예측값을 사용하여 정확도 계산
accuracy = accuracy_score(target, pred)

# 3) 오류율 계산
error_rate = 1 - accuracy
print(round(error_rate, 3))
# 0.478
```
