```python
## 문제 조건 : 제공된 데이터(attrition.csv)는 직원의 나이, 연봉, 근무 특성 등에 따른 이직 여부(attrition)에 관한 자료이다.
## 이 데이터를 이용하여 로지스틱 회귀분석을 수행하고, 답하시오
## 제공 데이터 : 종속변수 attrition(이직 여부/ 0=잔류, 1=이직) / 독립변수 age, income, overtime(0=아니오, 1=예)

# 문제 1. 로지스틱 회귀 모델을 적합한 후, p-value가 0.05 보다 작은 변수의 회귀 계수를 구하시오 (절편항 제외, 소수 셋째 반올림)
# 1. 라이브러리 및 데이터
import pandas as pd
from statsmodels.formula.api import logit
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/attrition.csv") ## 링크는? 복붙

## 로지회귀 적합
model = logit('attrition~age+income+overtime', data=df).fit()
print(model.summary()) # 통계 요약 확인

## 조건 : p-value가 0.05 보다 작은 변수 -> summary() 결과에서 구할 수 있음 "P>|z|" 영역 확인
## 절편(Intercept)항 제외이므로 p-value가 0.05보다 작은 변수는 'income'
# 회귀 계수 구하기 (params) ** 상관계수 : corr()
value = model.params['income'] ## param에 변수 지정할 때 () 아니고 []임. 인덱스 형식을 따름!
print(round(value, 3)) # 소수 셋째 반올림 : -0.002
# 답안 : -0.002


## 문제 2. 나이(age)가 1 증가할 때 이직할 오즈비(Odds Ratio)를 구하시오 (소수 셋째까지 반올림)
# 오즈비다? numpy
import numpy as np
# odds_ratio = np.exp(df['age'])  // 아님
odds_ratio = np.exp(model.params['age']) ## 회귀계수에 자연상수 e를 취한 값이 오즈비 이므로 회귀계수 model.params에 변수 ['age']를 적용함
print(round(odds_ratio,3)) # 근데 이제 반올림
# 결과 및 답안 : 0.996
### ********* 로지스틱 회귀에서 오즈비를 구한다 ? 회귀계수를 np.exp() 수행 **************** 회.귀.계.수 라고


## 문제 3. 새로운 직원(age=40, income=4,500, overtime=1)에 대해, 모델을 이용하여 이직 확률을 예측하시오 (소수 셋째 반올림)
## 새 데이터 예측을 해라? 데이터 프레임 생성 후 predict - 아까 쓴 로지 모델 쓰셈
new_data = pd.DataFrame({
    'age' : [40],
    'income' : [4500],
    'overtime' : [1]  ## 데이터 프레임 데이터는 : 로 지정함 (= 아님 ㅠ)
})

# 새 데이터로 로지 모델
# new_model = logit('attrition~age+income+overtime', data=new_data).fit()
# print(new_model.summary())  ### 놀랍게도 틀림

## 모델은 아까 그거 쓰고, 데이터만 새로운 데이터로 예측 수행하는거임
pred = model.predict(new_data) ## 여기 기존 model에 .predict() 함수 인자로 new_data를 넣으라는 것임
print(pred[0]) ## 인덱스 번호까지 출력되니까 [0] 붙여서 값만 가져오기. 결과 : 0.6969167516222728
print(round(pred[0], 3)) # 소수 셋째 반올림
# 답안 : 0.697
## 두 번째 문제는 아래 쉘에
```

```python
### 아니 무슨 문제가 또 있음
### 문제 조건 : 제공된 데이터(heating_load.csv)는 건물 특성과 난방 부하(heating_load)에 관한 자료이다. 다중회귀 분석을 수행하고, 답하시오
### 제공 데이터 : 종속변수 heating_load(난방 부하) / 독립변수 wall, roof, glazing, height

## 문제 1. 모든 독립 변수를 포함한 회귀 모형을 적합하시오. 이때 절편을 제외한 회귀계수의 합을 구하시오.(소수 셋째 반올림)
# 다중 회귀면 ols() 이였던가.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/heating.csv")  ## 배고프다. 밥.  먹었는데.
# 데이터 확인
print(df.head(3))

# 다중회귀 적합
from statsmodels.formula.api import ols
model = ols('heating_load~wall+roof+glazing+height', data=df).fit()
# print(model.summary()) # 통계 요약 확인

# 절편(Intercept) 제외 회귀계수 합 -> 회귀계수 model.params / 합 sum() / 절편 제외 [1:] / 는 아는데 어떻게 조합하지
value = model.params[1:].sum() ## 정답을 보고나면 쉽더라. 저거 그냥 갖다 붙여서 조합하면 됨.
print(round(value, 3)) # 소수 셋째 반올림. 결과 : 0.254
# 답안 : 0.254


## 문제 2. 유의한 변수(유의수준 0.05 이하) 만을 사용하여 다중회귀분석을 다시 수행, R^2(결정계수) 값? (소수 셋째 반올림)
# 유의수준은 summary() 결과 보고 확인하면 되겠고. ols(), R^2 결정계수가, rsqured 였던가.
# 유의한 변수(P<|z| 0.05 이하) : 어... ? 아. roof, blazing, height (Intercept 제외)
# 다중회귀 재수행
model = ols('heating_load~roof+glazing+height', data=df).fit()
print(model.summary()) # 통계 요약 확인

# 결정계수 (R^2)
print(model.rsquared) # 결과 : 0.7536194823637614
print(round(model.rsquared, 3)) # 소수 셋째 반올림 : 0.754
# 답안 : 0.754     // 회귀계수 함수는 모르더니 결정계수를 알고있었네. 굿👍


## 문제 3. 새로운 건물(wall=20, roof=150, glazing=20, height=5)이 주어졌을 때, 위 문제에서 사용한 모델을 이용하여
## 난방부하(heating_load) 예측 (소수 셋째 반올림)
# 새 데이터? 새 데이터 프레임 생성. / 재적합 하지 말고 모델 예측 함수 인자값만 새 데이터로.
new_data = pd.DataFrame({
    'wall':[20],
    'roof':[150],
    'glazing':[20],
    'height':[5]  ## 변수 한 줄이라도 빼먹으면 에러남
})
pred = model.predict(new_data)
print(round(pred[0], 3)) ## [0] 붙여서 값만 보기 + 소수 셋째 반올림
# 결과 : 79.612

## 문제 하나하나 푸는데 시간 실화냐. 내일까지 빠듯하겠는데.
```

### 문제1-1

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/attrition.csv")

# 1) 로지스틱 회귀분석 수행
from statsmodels.formula.api import logit
model = logit("attrition ~ age + income + overtime", data=df).fit()

# 2) 회귀 분석 결과 출력
print(model.summary())

# 3) income 회귀계수 출력
print(round(model.params["income"], 3))

# 정답: -0.002
```

### 문제1-2

```python
# 1) 오즈비 계산
import numpy as np
result = np.exp(model.params["age"])

# 2) 반올림 후 결과 출력
print(round(result, 3))

# 정답: 0.996
```

### 문제1-3

```python
# 1) 새로운 직원 데이터 만들기
new = pd.DataFrame({"age":[40], "income":[4500], "overtime":[1]})

# 2) 예측
pred = model.predict(new)

# 3) 반올림 후 결과 출력
print(round(pred[0], 3))
```

### 문제2-1

```python
import pandas as pd

# 1) 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/heating.csv")

# 2) 다중선형회귀분석
from statsmodels.formula.api import ols
model = ols("heating_load ~ wall + roof + glazing + height", data=df).fit()

# 3) 회귀 분석 결과 출력
print(model.summary())

# 4) 절편 제외 회귀계수들의 합 구하기
result = model.params[1:].sum()
print(round(result, 3))

# 정답: 0.254
```

### 문제2-2

```python
# 1) 다시 회귀 분석
model = ols("heating_load ~ roof + glazing + height", data=df).fit()

# 2) 결정계수(R-squared) 확인
print(model.summary())

# 정답: 0.754
```
