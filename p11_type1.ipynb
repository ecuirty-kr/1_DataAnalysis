```python
### 문제 조건 : 제공된 데이터(greenhouse_gas.csv)는 2001년부터 2021 년까지 30개 국가의 연도별 온실가스 배출량 데이터일 때, ,답하시오
## 문제 1. 2001 년부터 2021년까지 각 연도별로 온실가스 배출량이 가장 많은 국가를 찾고, 그 중 1위를 가장 많이 차지한 국가의 1위 횟수 ? (정수형)
# 1. 라이브러리 &  데이터
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/greenhouse_gas.csv')
# 데이터 확인
# print(df.head(4))

# 조건 : 2001년부터 2021년까지 / 각 연도별 / 배출량 가장 많은 국가 / 1위 가장 많은 국가 1위 횟수
# 날짜 데이터 datetime / df['Country'] / idxmax() / ... 아 기본기 부족
# print(df.columns) # 모든 컬럼명 확인

## 모범 답안
# 1. 각 연도별 - 연도 컬럼만 지정
cols = df.columns[1:] ## 컬럼[0] : Country
# print(cols) ## 연도만 담겼는지 확인

# 2. 각 연도별 최댓값의 행 번호 : 배출량 가장 많은
idx = df[cols].idxmax()
# print(idx) # ??  어 근데 위 코드는 맞음

# 3. 배출량 최다 인덱스의 데이터 프레임
df2 = df.loc[idx] # loc : 지정 행번호 탐색
# print(df2) ## 아니 출력 값들이 다 이상함

# 4. 국가별 1위 횟수 합계
print(df2['Country'].value_counts())
# 결과 : China     11     /     Brazil    10
# 답안 : 11


# ******* 다른 방법
# 지정 인덱스 주소에 있는 데이터 프레임의 국가명
df3 = df.loc[idx, 'Country']

# 국가별 1위 횟수
# print(df2.value_counts())  ## China, Brazil의 모든 연도별 배출량 확인 가능 - 마지막 열에 총합 있음

## 문제 2. 다음 공식으로 각 국가의 CAGR(연평균 성장률)을 계산하고, 그 중 최댓값을 소수점 셋째 자리까지 구하시오
## CAGR = {(EV/BV)^1/n} - 1    * EV : 최종 값(2021년 배출량) / BV: 초기 값(2001년 배출량) / n: 전체 기간 (2021년 - 2001년 = 20년)

# 문제 요구사항 대로 식 대입
df['CAGR'] = (df['2021']/df['2001']) ** (1/20) - 1 ## 수식보고 쫄지말고 값 하나씩 대입부터 해보라
## 위에 제곱하는 값(1/20)은 꼭 괄호로 묶어줘야 함. 안 하면 음수값 나옴

# 최댓값
value = df['CAGR'].max()
print(round(value, 3)) # 결과 : 0.047
# 답안 : 0.047

## 1 유형도 문제가 2개인, 날개 문제도 2개 이상씩인.
```

```python
### 문제 조건 : 주어진 데이터에서 다음 작업을 순서대로 수행하시오
### 제공 데이터 : sensor_data.csv

## 문제 1. 결측치가 가장 많은 컬럼을 찾으시오
# 1. 라이브러리 및 데이터 확인
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/sensor_data.csv')
print(df.head(4))
# 결측치
print(df.isnull().sum()) # 직접 확인 : S2_Temp  ..idxmax 써도 되나
print("최다 결측치 : ", df.isnull().sum().idxmax()) # 최다 결측치 :  S2_Temp   // 오 맞춤

## 문제 2. 해당 컬럼의 결측치를 중앙값으로 대체하시오
# 1. 해당 컬럼 (S2_temp) / 결측치 대체 : .replace(이전 값, 새 값) / 중앙값 median()
middle = df['S2_Temp'].median()
# null = df['S2_Temp'].isnull() ## ???? 결측치만 어떻게 빼오지
# df['S2_Temp'] = df['S2_Temp'].replace(null, middle)

## .replace() 가 아니라 .fillna()로 결측치만 바로 대체 가능 - 왜 까묵었니
df['S2_Temp'] = df['S2_Temp'].fillna(middle)
print(df['S2_Temp'].isnull().sum())  ## 결측치 개수 : 0 확인

## 문제 3. 결측치 대체한 후 해당 컬럼의 평균을 소수점 셋째 자리까지 구하시오
print("평균 : ", round(df['S2_Temp'].mean(), 3)) # 평균 :  25.055
# 답안 : 25.055

### 왜 문제가 하나 더 있는지 설명 좀
```

```python
### 문제 조건 : 제공된 데이터(order_data.csv)는 온라인 쇼핑몰의 주문 데이터. 전처리 후 답하시오.
### 전처리 수행 : 새로운 컬럼 order_amt 생성 : order_amt = quantity * price
### 새로운 컬럼 cancel_TF 생성 : 주문번호가 'C'로 시작하면 True, 아니면 False ('C'는 취소 주문)
### cancel_TF 가 'T'인 행의 order_amt를 음수로 변환

## 문제 1. 취소 주문 중 order_amt의 절댓값이 가장 큰 주문번호를 구하시오.
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/order_data.csv')
print(df.head(4)) # 데이터 확인

# 2. 전처리
## 모범 답안
# order_amt 새 컬럼 생성
df['order_amt'] = df['quantity'] * df['price']
print(df.head(3)) # 생성 확인

# cancel_TF 생성 : 주문 번호가 'C' 로 시작하면- (주문번호 컬럼명 : order_no)
df['cancel_TF'] = df['order_no'].str.contains('C')  ## str.contains('값') : 해당 값이 있는지 확인하여 boolean 값으로 출력 (있으면 True, 없으면 False)
## startswith('값') 으로 사용 가능 : df['cancel_TF'] = df['order_no'].str.startswith('C')  // 이게 더 정확하긴 함.
print(df.head(3)) # 생성 확인 * 값도 확인할 것

# cancel_TF 가 'T'인 행의 order_amt를 음수로 변환 (???)
cond = df['cancel_TF'] == True  ## 데이터 확인해보면 'T'가 아니라 'True' 값임. 고려해서 조건문 작성
df.loc[cond, 'order_amt'] = df.loc[cond, 'order_amt'] * -1 ## 음수 변환 : 함수 아니고 * -1 로 음수값으로 계산 수행
## loc[cond, '컬럼명'] : cond 조건에 맞는 '컬럼'의 행 지정
print(df.head(5)) ## 왜 음수로 변환이 안 됨 -> cond 조건문에 True 를 따옴표로 묶어서 인식이 안 됐음 (해결)

## 뭔 전처리가 문제보다 어려움

## 문제 1. 취소 주문 중 order_amt 의 절댓값이 가장 큰 주문번호 ?
# 취소 주문 : cancel_TF = True / 절댓값 : .abs() / 가장 큰 주문번호 : ['order_no'].idxmax()
cond2 = df['cancel_TF'] == True # 조건 : 취소 주문
print(df.loc[cond2, 'order_amt'].abs()) ## 이렇게 하면, 취소 주문에 해당하는 'order_amt' 컬럼의 절댓값이 출력되나 -> 맞음
index = df.loc[cond2, 'order_amt'].abs().idxmax() # 취소주문 'order_amt' 컬럼의 최대 절댓값 인덱스
df.loc[index, 'order_no'] # loc이 만능이네. 결과 : C542426  (과연!)
# 답안 : C542426  (오 맞음. 굿)

## 문제 2. 고객ID(cus_id)별로 order_amt를 합산하여 가장 큰 값을 소수점 둘째 자리까지 구하시오.
# 고객ID별 df['cus_id'] / ['order_amt'].sum() 합산 / 가장 큰 값 idxmax() - 반복문..? groupby(?)

## 모범 답안 : groupby() 맞음. DB 인줄
# 고객별 order_amt 합산
df = df.groupby('cus_id')['order_amt'].sum()

# 합계 중 최댓값
print(df.max()) # 결과 : 2877.21   * idxmax() 아니고 max()임
# 답안 : 2877.21

## groupby가 파이썬에도 존재함. 그룹"별" 하면 groupby('변수') + ['컬럼'] 도 가능
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/sensor_data.csv')

# 1) 각 컬럼의 결측치 개수 확인
print(df.isnull().sum())

# 2) 결측치가 가장 많은 컬럼 찾기
col = df.isnull().sum().idxmax()
# print(col) #S2_Temp

# 3) 해당 컬럼의 중앙값 계산
m = df[col].median()
# print(m) #25.0

# 4) 결측치를 중앙값으로 대체
df[col] = df[col].fillna(m)

# 5) 평균 계산 (소수점 셋째 자리)
print(round(df[col].mean(), 3))
```

```python
# 전처리 수행
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/order_data.csv')

# 1) order_amt 생성
df['order_amt'] = df['quantity'] * df['price']

# 2) cancel_TF 생성
df['is_cancel'] = df['order_no'].str.contains('C')     # True/False 반환 - 방법1
#df['is_cancel'] = df['order_no'].str.startswith('C')  # True/False 반환 - 방법2

# 3) 취소 주문을 음수로 변환
cond = df['is_cancel'] == True
df.loc[cond, 'order_amt'] = df.loc[cond, 'order_amt'] * -1
df.head()
```

```python
# 문제 3-1
# 1. 취소 주문만 필터링
cond = df['is_cancel'] == True
cancel_df = df[cond]

# 2. order_amt 절댓값이 가장 큰 행의 인덱스
im = cancel_df['order_amt'].abs().idxmax()

# 3. 특정 인덱스의 전체 행 출력 (모든 컬럼)
print(cancel_df.loc[im])
# 정답: C542426
```

```python
# 문제 3-2
# 1. 고객별 order_amt 합계 계산
df = df.groupby('cus_id')['order_amt'].sum()

# 2. 합계 중 최댓값
ma = df.max()

# 3. 소수점 둘째 자리까지 출력
print(round(ma, 2))
# 정답: 2877.21
```
