```python
### 문제 조건 : 로지스틱 회귀 / data: churn.csv
### 뭐야 문제가 이게 끝임

## 문제 1. 주어진 데이터에서 로지스틱 회귀 분석을 수행해 유의확률(p-value)이 0.05 이상인 유의하지 않은 독립변수의 개수를 구하시오
## 종속변수 : Churn / 독립변수 : Churn을 제외한 모든 변수
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn.csv")
# 데이터 확인 - 잊지 말고!
print(df.head(4))
print(df.columns) ## ******** 모든 컬럼명 확인 *************** (컬럼 수가 너무 많으면 중간에 생략됨)

# 로지 회귀
from statsmodels.formula.api import logit
model = logit('Churn~AccountWeeks+ContractRenewal+DataPlan+DataUsage+CustServCalls+DayMins+DayCalls+MonthlyCharge+OverageFee+RoamMins', data=df).fit()
# print(model.summary()) ## 통계 요약 p-value > 0.05 확인 : AccountWeeks, ContractRenewal, MonthlyCharge, OverageFee,  RoamMins ,,... 너무 많은디
print(sum(model.pvalues > 0.05)) # 결과 : 8
# 답안 : 8

## 문제 2. 유의확률 0.05 미만인 유의한 변수만을 사용해 다시 로지스틱 회귀 분석을 수행하시오. 이 회귀식의 유의한 회귀 계수(상수항 포함) 합계? (반올림 소수 셋째)
# 유의확률 0.05 미만 데이터 필터링
cond = model.pvalues < 0.05
# print(df[cond]) ## 필터링ㅜ 어렵다 뭐지 이거
# -> 그냥 통계 요약 보고 0.05 미만 데이터 직접 입력
# 로지 회귀
from statsmodels.formula.api import logit
model = logit('Churn~DataUsage+DayMins', data=df).fit()
print(model.summary()) ## 통계 요약
print(round(sum(model.params), 3)) # 아 상관계수 외우고 회귀계수 잊어버림 (회귀계수 params / 상관계수 corr())
# 결과 및 답안 : -1.213

## 이 회귀식의 유의한 회귀 계수(상수항 포함) 합계 = 상수항을 포함한 회귀계수(params)의 합계(sum())


## 문제 3. 위에서 수행한 로지스틱 회귀식에서 'DataUsage' 변수가 5 만큼 증가할 때 오즈비(Odds Ratio)는? (소수 셋째 반올림)
# 오케. 오즈비 numpy
import numpy as np
# result = np.exp(df['DataUsage']**5) ## 인자값에서 또 버벅
result = np.exp(model.params['DataUsage']*5)  ## np.exp() 인자값으로 "로지스틱 회귀식에서 '변수'"일 경우 model.params['변수']로 들어감 / params 용도 숙지 좀!!! 회귀 계수 + 오즈비 함수 인자
print(round(result, 3)) # 결과 : 0.428
# 답안 : 0.428


### 문제 2. 다중 선형 회귀 / data: piq.csv
# ### 이전 문제라서 문제 설명이 많이 없나보네

# ## 문제 1. 주어진 데이터를 이용해 종속변수(PIQ)와 독립변수(Brain, Height, Weight)로 다중 선형 회귀 분석을 수행하시오
# ## 이 때 유의확률(p-value)이 가장 작은 변수의 회귀 계수 값을 구하시오 (소수 셋째 반올림)
# # 1. 라이브러리 및 데이터
# import pandas as pd
# df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/piq.csv") # 링크는 아래서 복.붙
# # 데이터 확인
# print(df.head(5))

# ## 아래 쉘에서 해야겠다.
```

```python
## 문제 2. 다중 선형 회귀 / data: piq.csv
### 이전 문제라서 문제 설명이 많이 없나보네

## 문제 1. 주어진 데이터를 이용해 종속변수(PIQ)와 독립변수(Brain, Height, Weight)로 다중 선형 회귀 분석을 수행하시오
## 이 때 유의확률(p-value)이 가장 작은 변수의 회귀 계수 값을 구하시오 (소수 셋째 반올림)
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/piq.csv") # 링크는 아래서 복.붙
# 데이터 확인
print(df.head(5))

# 2. 다중 선형 회귀 분석
from statsmodels.formula.api import ols
model = ols('PIQ~Brain+Height+Weight', data=df).fit()
print(model.summary()) # p-value(P>|t| 값 확인) 가장 작은변수 : Brain
print(round(model.params['Brain'], 3)) # 회귀 계수 값 / 결과 : 2.343
# 답안 : 2.343

## 문제 2. 위 문제에서 적합한 모델의 결정계수(R-squared) 값을 구하시오 (소수 둘째 반올림)
# R 결정계수 = rsquared
print(round(model.rsquared, 2)) # 결과 : 0.37
# 답안 : 0.37

## 문제 3. 뇌 크기(Brain)가 90, 키(Height)가 70, 몸무게(Weight)가 150일 때의 PIQ 예측하시오 (정수로 반올림)
# 새로운 값 + 예측 / DataFrame, predict()
new_data = pd.DataFrame({
    'Brain' : [90],
    'Height' : [70],
    'Weight' : [150]
})
pred = model.predict(new_data) # 기존 모델에 새 데이터로 예측
print(round(pred[0])) # 정수 반올림 인수는 없는건가
# 결과 및 데이터 : 106

## 이번 합격 못하면 다음은 없을 듯. 시간이. 하지만 피곤타.
```

### 문제1-1

```python
import pandas as pd
# df = pd.read_csv("churn.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn.csv")

# 1) 로지스틱 회귀분석을 위한 포뮬라 생성
formula = "Churn ~ AccountWeeks + ContractRenewal + DataPlan + DataUsage + CustServCalls + DayMins + DayCalls + MonthlyCharge + OverageFee + RoamMins"

# 2) 로지스틱 회귀 모델 생성 및 학습
from statsmodels.formula.api import logit
model = logit(formula, data=df).fit()

# 3) 유의하지 않은 독립변수의 개수 구하기
print(model.summary())
sum(model.pvalues[1:] > 0.05) # Intercept는 독립변수가 아니므로 제외
```

### 문제1-2

```python
# 1) 유의한 변수 선택
print(model.pvalues < 0.05)
formula = "Churn ~ DataUsage + DayMins"

# 2) 로지스틱 회귀
model = logit(formula, data=df).fit()

# 3) 유의한 회귀계수 합계
print(model.summary())
round(sum(model.params),3)
```

### 문제1-3

```python
# 1) 변수의 회귀계수 추출
import numpy as np
coef = model.params['DataUsage']

# 2) 오즈비 계산
round(np.exp(coef * 5),3)
```
