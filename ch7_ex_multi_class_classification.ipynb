```python
# 1. 문제 정의
## 평가 지표: f1-macro
## 예측 컬럼 : Credit_Score(Good, Standard, Poor) 세 가지 값으로 예측
## 제출 파일명 : "result.csv" (pred : 예측값 컬럼 1개)

# 2. 라이브러리 불러오기 + 데이터 가져오기
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch7/score_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch7/score_test.csv")

# 3. 탐색적 데이터 분석 (EDA) - 데이터 크기, 자료형, 데이터 샘플(생략), 결측치, target 빈도 수
# print("\n\n------- 데이터 크기 -----------")
# print(train.shape, test.shape) ## (4198, 21) (1499, 20)

# print("\n\n------- 데이터 자료형 -----------")
# print(train.info(), test.info())  ## float 16 / object 4(train은 4개 / test는 3개 - Credit_Score 컬럼 없음)

# print("\n\n------- 데이터 결측치 -----------")
# print(train.isnull().sum(), test.isnull().sum())  ## 결측치 없음

# print("\n\n------- Target 빈도 수 -----------")
# print(train['Credit_Score'].value_counts()) ## Standard  2225 / Poor 1232 / Good  741

# 4. 데이터 전처리
## 원-핫 인코딩 시 target 컬럼 제외 (자료형이 object 형이라 컬럼이 다수 개 생성됨)
target = train.pop('Credit_Score')
train = pd.get_dummies(train)
test = pd.get_dummies(test)

## 5. 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0) ## 오 여기까지 수월

## + 분할된 데이터 크기 확인
print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)  ## y_tr, y_val 값이 시리즈 형태(값, )로 존재하는지 확인

## 6. 머신러닝 학습 및 평가
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)  ## 랜덤 포레스트 사용 (random=0 고정값)
rf.fit(X_train, y_train) ## 학습
pred = rf.predict(X_val) ## 예측 X_val : f1_score 평가지표 사용하므로 predict()

## 평가 지표 f1_score
from sklearn.metrics import f1_score
score = f1_score(y_val, pred, average='macro') ## f1-macro 평가지표 사용 시 옵션 값 average='macro' 추가
print("f1-macro  : ", score) ## f1-macro 평가 값 확인 0.70045...(이하 생략)

## 7. 예측 및 결과 파일 생성
pred = rf.predict(test) ## 실제 데이터로 예측 수행
submit = pd.DataFrame({'pred':pred}) ## 데이터 프레임 생성
submit.to_csv("result.csv", index=False) ## 제출 파일 생성

## 결과 파일 검증
file = pd.read_csv("result.csv")
display(file) ## 검증 완료
```
