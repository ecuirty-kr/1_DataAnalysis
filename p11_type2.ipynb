```python
## 문제 정의 : 제공된 학습용 데이터(train.csv)는 온라인 게임 플레이어의 활동 데이터이다. 학습용 데이터를 활용하여
## 유저 등급(User Level)을 예측하는 모형을 개발하고, 모형을 평가용 데이터(test.csv)에 적용하여 유저 등급을 예측하시오
## 예측 결과를 아래의 [유의 사항]과 [제출 형식]울 준수하여 CSV(result.csv)을 생성하는 코드를 제출

## 예측 결과는 F1 Score(macro) 평가 지표에 따라 평가
## 타겟 변수(User_Level)는 0, 1, 2, 총 3개의 클래스로 구성됨
## 제출 파일명 : result.csv
## 예측 유저 등급 컬럼명 : pred
## 제출 컬럼 개수 : pred 컬럼 1개
## 평가용 데이터 개수와 예측 결과 데이터 개수 일치 : 903개

## 제공 데이터 : train.csv 2,614개 / test.csv 903개

# 1. 라이브러리 및 데이터
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/game_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/game_test.csv")
# 데이터 확인
# print(train, test)

# 2. 탐색적 데이터 분석 (데이터 크기, 자료형, 결측치, target 빈도 수)
print(train.shape, test.shape)  ## (2614, 10) (903, 9)
# print(train.info(), test.info())  ## float 4 / int 4  : train 1 int 추가 (all 수치형)
# print(train.isnull().sum(), test.isnull().sum())  ## 결측치 없음
print(train['User_Level'].value_counts()) ## 다중 / 빈도 수 : 1  : 1313 |  0  : 1165  | 2  :  136

# 3. 데이터 전처리 - 모두 수치형 / 결측치 없음 -> 크게 작업할 것 없음
target = train.pop('User_Level')

# 4. 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

## 오 다음 까먹음. 머신러닝 학습이었던가
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
pred = rf.predict(X_val)

################################################################### 여기까진 괜찮았음

# 6. 모델 평가 (평가 지표 : f1-score)
from sklearn.metrics import f1_score
f1 = f1_score(y_val, pred, average='macro')  ## 이거 까먹지 마라고ㅜ f1_score(y_val, pred, average='macro')
print("f1-score 확인 : ", f1)  ## 결과 : 0.7241845968568211

## (!New!) 6-1. LightGBM  - LightBGM 아님
# 위에 랜덤 포레스트 모델 만드는 것과 유사함
import lightgbm as lgb  # 라이브러리 (얘는 import ~ as)

lg = lgb.LGBMClassifier(random_state=0, verbose=-1) ## 얘도 값 고정(random_state=0)은 동일 / vebose=-1 옵션 들어감
lg.fit(X_train, y_train)

pred = lg.predict(X_val)  ## 얘 인자값은 또 왜 생각 안 남 - 랜포랑 같다고.!.!
print("LGBMClassifier : ", f1_score(y_val, pred, average='macro')) ## 결과 : 0.7989718332478102

# 두 모델 평가 결과 rf < lg 이므로 LightGBM을 사용하도록
# 7. 예측 및 최종 제출 파일 생성
pred = lg.predict(test) # test 데이터가 진짜임
## 파일 생성하는거 형식 헷갈리지 말기
result = pd.DataFrame({'pred':pred}) ## 데이터 프레임 생성하고.
result.to_csv("result.csv", index=False) ## 기존 인덱스 없이 "result.csv" 파일로 to_csv 수행


## 검증 방법 알아두기
print(pred.shape)  ## (903,) : test 행의 수 = 903

file = pd.read_csv("result.csv")
file ## 파일 생성 확인
```

```python
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/game_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/game_test.csv")

# 데이터 전처리
target = train.pop('User_Level')

# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

# 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)

# macro F1 score
from sklearn.metrics import f1_score
print("랜덤포레스트")
print(f1_score(y_val, pred, average='macro'))

# LightGBM
import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state=0, verbose=-1)
lg.fit(X_tr, y_tr)
pred = lg.predict(X_val)
print("lightgbm")
print(f1_score(y_val, pred, average='macro'))

# 최종 제출 파일 (lightGBM)
pred = lg.predict(test)
result = pd.DataFrame({
    'pred':pred
})
result.to_csv("result.csv", index=False)
```

```python
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/game_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/game_test.csv")

# 데이터 전처리
target = train.pop('User_Level')

# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0, stratify=target)

# 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0, class_weight='balanced')
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)

# macro F1 score
from sklearn.metrics import f1_score
print("랜덤포레스트")
print(f1_score(y_val, pred, average='macro'))

# LightGBM
import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state=0, verbose=-1, is_unbalance=True)

lg.fit(X_tr, y_tr)
pred = lg.predict(X_val)
print("lightgbm")
print(f1_score(y_val, pred, average='macro'))

# 최종 제출 파일 (lightGBM)
pred = lg.predict(test)
result = pd.DataFrame({
    'pred':pred
})
result.to_csv("result.csv", index=False)
```

```python
# LightGBM - 전체 데이터로 학습
lg = lgb.LGBMClassifier(random_state=0, verbose=-1, is_unbalance=True)
lg.fit(train, target)

# 최종 제출 파일 (lightGBM)
pred = lg.predict(test)
result = pd.DataFrame({
    'pred': pred
})
result.to_csv("result.csv", index=False)
```
