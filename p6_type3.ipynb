```python
##### D-1 운 좀 따라줘라
#### 문제 조건 : 감기약 위약(플라시보)을 투여받은 환자의 부작용은 항암약 위약(플라시보)을 주여받은 환자의 부작용과 차이가 있는가?
#### 효과가 없는 위약(가짜약)을 환자에게 제안하지만, 환자의 심리적인 요인에 의해 병세가 호전되기도 하고 부작용을 느끼기도 한다.
#### 감기약(위약)을 투여받은 환자들 중 일부는 부작용을 경험했는데, 이 부작용의 알려진 발생 비율은 다음과 같다
#### 구분  1 / 아픔 / 10%     구분 2 / 조금 아픔 / 5%     구분 3 / 속쓰림 / 15%     구분 4 / 이상 없음 / 70%
#### 항암약(위약)을 투여받은 셀제 환자 그룹에서 관찰된 부작용은 다음과 같다. (데이터 파일 확인)
#### 귀무가설 : 감기약(위약)의 부작용과 항암약(위약)의 부작용은 동일하다 / 대립가설 : 감기약(위약)의 부작용과 항암약(위약)의 부작용은 다르다

## 문제 1. 항암약을 투여받은 환자 중 '이상 없음'의 비율을 0 과 1 사이로 나타내시오
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.DataFrame({
    "항암약":[4,4,3,4,1,4,1,4,1,4,4,2,1,4,2,3,2,4,4,4]
})
print(df.head()) # 데이터 확인

# 빈도 계산 : df의 구하려는 컬럼['항암약']에 해당하는 조건식을 제시(==4) / 조건에 맞는 컬럼의 개수 sum() 함수 사용
cnt = sum(df['항암약']==4) # 조건에 부합하는 컬럼 수 : 11
print(cnt/len(df)) # 비율을 구해야 하므로 '총합/데이터 크기'  연산 수행
# 결과 및 답안 : 0.55

## value_counts() 함수 사용하는 방법도 존재
print(df['항암약'].value_counts(normalize=True)) ## 모든 변수에 대해 비율값 출력 확인 (4 에 해당하는 값 확인)


## 문제 2. 감기약의 예상 부작용 비율과 항암약의 부작용 관찰값이 통계적으로 유의미하게 차이가 있는지 확인하려 한다
## 카이제곱 검정을 사용하여 검정 통계량을 구하시오
# 카이제곱 검정 : 적합성 검정 [라이브러리 scipy.stats import chisquare]
from scipy.stats import chisquare
# 각 카테고리의 비율을 리스트로 생성
prob = [0.1, 0.05, 0.15, 0.7] # 표에 제시된 비율 : 10%  5% 15% 70%

# 기대 빈도 수 계산 "비율*총 데이터 수"
print("총 데이터 수 : ", len(df)) # 총 데이터 수 :  20
expected_counts = [0.1*20, 0.05*20, 0.15*20, 0.7*20] # 각 비율에 총 데이터 수(20)을 곱한 값들을 리스트로 생성
print(expected_counts) # 결과 : [2.0, 1.0, 3.0, 14.0]

# 관찰 빈도 수 계산
observed_counts = df['항암약'].value_counts().sort_index().to_list() ## '항암약' 데이터 개수를 정렬하여 리스트로 생성
print(observed_counts) # 결과 : [4, 3, 2, 11]

# 카이제곱 검정 수행
print(chisquare(f_obs=observed_counts, f_exp=expected_counts)) # chisquare(class1(관찰 빈도 수), class2(기대 빈도 수))
# 결과 : Power_divergenceResult(statistic=np.float64(6.976190476190476), pvalue=np.float64(0.07266054733847573))
# 답안 : 6.976190476190476

## 문제 3. 위의 p-value 값을 구하시오
# 카이제곱 검정 결과에서 pvalue=np.float64(0.07266054733847573)
# 답안 : 0.07266054733847573

### 카이제곱 검정 공부 다시 해야겠다. 응용 좀.
### 문제 하나 더 있음
```

```python
#### 문제 조건 : 다중 선형 회귀 모델을 구축하고, 문제에 답하시오
#### data : data6-3-2.csv
#### 독립변수 : solar(태양 에너지), wind(바람의 세기), O3(오존 농도) / 종속변수 : temperature(온도)

## 문제 1. 주어진 데이터를 기반으로 다중 선형 회귀 모델을 구축하고, 독립변수 O3의 회귀 계수를 구하시오
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch6/data6-3-2.csv")
print(df.head(4)) # 데이터 확인

# 다중 선형 회귀 : ols(statsmodels.formula.api) / 회귀 계수(params)
from statsmodels.formula.api import ols
model = ols('temperature~solar+wind+o3', data=df).fit()
print(model.summary()) # 통계 요약 확인

# O3 회귀계수
print(model.params['o3']) # 결과 : 0.0749385437813658  // 왜 반올림 조건이 없지
# 답안 : 0.0749385437813658
### 모범 답안 : 0.0749 (summary() 에서 확인한 값 입력)

## 문제 2. 위에서 적합한 모델에서 'solar'와 'o3'이 고정된 상태에서 'wind'의 세기가 증가함에 따라 'temperature'가 감소한다
## 이 회귀 모델에서 'wind' 회귀 게수에 대한 p-value 값 ?
# 문제 말꼬기 넘어가지 말 것. 문제에서 요구하는 정보는 ['wind']의 회귀 계수임.!
print("p-value : ", model.pvalues['wind']) # 결과 : 0.7797177202071661
# 답안 : 0.7797177202071661

## 문제 3. 위에서 적합한 모델에서 solar:100, wind: 5, o3: 30 일 때, 예측값 ?
# 예측값 = model.predict() 문제 / 새 데이터 주어짐 -> 새 데이터 프레임 생성
new_data = pd.DataFrame({
    'solar' : [100],
    'wind' : [5],
    'o3' : [30]
})
pred = model.predict(new_data)
print(pred[0]) ## pred 출력 후 인덱스 같이 출력시 [0]로 값만 출력할 수 있게.
# 결과 및 답안 : 21.561630469479677

## 선형 / 로지 모델은 어느정도 되는거 같은데 카이제곱이 좀ㅁ...
```

### 문제1

```python
import pandas as pd
df = pd.DataFrame({
    "항암약":[4,4,3,4,1,4,1,4,1,4,4,2,1,4,2,3,2,4,4,4]
    })
print(df.head(3))
```

```python
# 1) 이상 없음(4)의 빈도 계산
cnt = sum(df['항암약']==4)

# 2) 항암약을 투여 받은 환자 중 '이상 없음' 비율 계산
ratio = cnt / len(df)
print(ratio)
```

```python
# 1) 카테고리별 비율 계산
print(df['항암약'].value_counts(normalize=True))
```

```python
from scipy.stats import chisquare

# 1) 각 카테고리의 비율을 리스트로 만들기
prob = [0.1, 0.05, 0.15, 0.7]

# 2-1) 기대 빈도수 계산
print("데이터 수: ", len(df))
expected_counts = [0.1*20, 0.05*20, 0.15*20, 0.7*20]
print(expected_counts)

# 2-2)기대 빈도수 계산 (다른 방법)
expected_counts = [x*len(df) for x in prob]
print(expected_counts)
```

```python
# 3) 관찰 빈도수 계산
observed_counts = df['항암약'].value_counts().sort_index().to_list()
print(observed_counts)
```

```python
# 4) 카이제곱 검정 수행
print(chisquare(f_obs=observed_counts, f_exp=expected_counts))
```

```python
# 참고 - f_obs, f_exp 생략 가능
print(chisquare(observed_counts, expected_counts))
```

### 문제2

```python
import pandas as pd
# df = pd.read_csv("data6-3-2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch6/data6-3-2.csv")

print(df.head(3))
```

```python
from statsmodels.formula.api import ols

# 1) R스타일 formula
formula = 'temperature ~ solar + wind + o3'

# 2) 회귀 모델 학습
model= ols(formula, data=df).fit()

# 3) 회귀 모델 요약 정보
summary = model.summary()
print(summary)
```

```python
# 2-1. o3에 대한 회귀계수
print("2-1. o3의 회귀계수:", model.params['o3'])
```

```python
# 2-2. wind에 대한 p-value
print("2-2. wind의 p-value:", model.pvalues['wind'])
```

```python
# 2-3. 예측값
# 1) 새 데이터를 데이터프레임으로 만들기
new_data = pd.DataFrame({
    'solar': [100],
    'wind': [5],
    'o3': [30]
})
print(new_data)
```

```python
# 2) 구축된 모델을 사용해 예측
pred = model.predict(new_data)
print(pred)
```
