# 2. 단일 표본검정

### 1. 기본학습

```python
## 단일 표본 검정
# 라이브러리 불러오기
import pandas as pd

# 데이터 프레임 생성 (가설 검정 수행할 데이터)
df = pd.DataFrame({
    'weights' :[122,122,120,119,125,115,121,118,117,127,
                123, 129, 119, 124, 114, 126, 122, 124, 121, 116,
           120, 123, 127, 118, 122, 117, 124, 125, 123, 121]
})
```

```python
## ## 단일 표본 검정 (t-test[t검정])
## scipy 라이브러리 활용 stats
from scipy import stats

t_statistic, p_value = stats.ttest_1samp(df['weights'], 120) ## stats로 .ttest_1samp() 불러옴 (단일 표본 검정이라 1samp임 = 하나의 샘플)
## ttest_1samp() 인자값 : 검정을 수행할 데이터 프레임, 가설 조건 "'weights'가 '120g 이다"
## t_statistic 받아온 값 확인
print("t-statistic : ", t_statistic) ## 결과 : 2.1644660352205833
## p_value 값 확인 : 0.05 이하일 때 대립 가설 채택
print("p-value : ", p_value) ## p-value :  0.03880039789894752

## p_value 값이 0.03 으로 0.05 보다 작은 값이므로 귀무가설을 기각하고, 대립가설을 채택함
# "팝콘 무게가 120g 가 아니다 라는 가설은 유의미한 통계"
```

# 3. 대응 표본검정

### 1. 기본학습

```python
## 대응 표본 검정 (한 집단에 대한 동일한 검정을 시간차를 두고 수행)
# 1. 라이브러리 불러옴
import pandas as pd
df = pd.DataFrame({
    'before' : [85, 90, 92, 88, 86, 89, 83, 87],
    'after' : [85.5,89.9,92.6,89.5,85.8,88.8,84.6,87.8]
})

# 2-1. 대응 표본 검정 수행 - scipy.stats의 ttest_rel() 사용
from scipy import stats
## 문제 조건 : 새 교육 프로그램 교육 전후의 성과 비교
# "교육 후 성과 증가" - 대립 가설 before < after 이어야 함 / "귀무 after > before"  -> 얘가 조건이 되므로 alternative='less'
t_statistic, p_value = stats.ttest_rel(df['before'], df['after'], alternative='less')  ## 함수 쓸 때 라이브러리 잊지말고 stats. 붙이시길
print("t-statistic : ", t_statistic) # 값 :-2.2127749675452324
print("p-value : ", p_value) # 값 : 0.03127028733756238

## p-value(0.0312) < 0.05 이므로 대립가설 채택 "교육이 성과에 대한 효과가 있다"

# 2-2. 단측 검정 (반대 상황)
from scipy import stats
## 귀무 : 새 교육은 효과가 없다 / 대립 : 효과가 있다 'after'-'before' 일 때, 값 > 0 성립해야 함
print(stats.ttest_rel(df['after'], df['before'], alternative='greater'))  ## 앞 DF - 뒤 DF 값 > 0 이어야 하므로 greater(값이 더 커야 함)
# 결과 TtestResult(statistic=np.float64(2.2127749675452324), pvalue=np.float64(0.03127028733756238), df=np.int64(7))

# p-value(0.0312) < 0.05 이므로 대립 채택 "새 교육 프로그램이 성과에 효과가 있다"

## 3. 양측 검정 (기본값 alternative 옵션 X)
print(stats.ttest_rel(df['after'], df['before'])) # 크다/작다 가 아닌 단순 차이가 발생한다의 개념
## 결과 : TtestResult(statistic=np.float64(2.2127749675452324), pvalue=np.float64(0.06254057467512476), df=np.int64(7))
# p-value(0.0625) > 0.05 이므로 귀무 채택 "교육 전후 차이가 없다"
```

 ### 2. 심화학습

```python
## 문제 조건 : 동일한 직원 그룹에 대해 교육 전과 후의 성과 비교 (대응 표본 검정 : ttest_rel())
import pandas as pd
## 데이터는 복붙
df = pd.DataFrame({
    'before':[85, 90, 92, 88, 86, 89, 83, 87],
    'after':[86, 92, 94, 89, 84, 90, 84, 88]
})

## 정규 분포를 따른다고 명시되지 않을 시 샤피로-윌크 검정 수행
from scipy import stats
## 샤피로-윌크 : 귀무="정규분포 따름" / 대립="정규분포 따르지 않음"
df['diff'] = df['after'] - df['before'] ## after - before 한 값을 구해서 shapiro() 수행
stats.shapiro(df['diff']) # 결과값 : ShapiroResult(statistic=np.float64(0.6886147375920879), pvalue=np.float64(0.0016734051223900109))

# p-value(0.0016) < 0.05 이므로 대립 채택 "정규 분포를 따르지 않음"
## 정규성을 만족하지 않으므로 윌콕슨 검정 수행
stats.wilcoxon(df['after'], df['before'], alternative='greater') ## 교육 후(after)의 값이 더 크다(greater)
## 결과값 : WilcoxonResult(statistic=np.float64(29.0), pvalue=np.float64(0.07421875))

# p-value(0.0742) > 0.05 이므로 귀무가설 채택 "새 교육 프로그램은 효과가 없다"
##  헷갈릴수도
```

# 4. 독립 표본검정

### 1. 기본학습

```python
## 독립 표본 검정 : 다른 두 집단의 평균값이 서로 다름을 판단하는 검정 "통계적으로 유의미하다"
# ttest_ind() 사용 a, b, alternative 인자 + equal_var(True/False) 추가됨

## 문제 조건 : 1, 2반의 평균 수학 시험 점수가 차이가 있는지 검정 (1, 2반 = 서로 다른 두 집단)
import pandas as pd
class1 = [85, 90, 92, 88, 86, 89, 83, 87]
class2 = [80, 82, 88, 85, 84]  ## 얘네는 데이터 프레임이 아닌 리스트 형태로 두 그룹 주어짐

from scipy import stats
print(stats.ttest_ind(class1, class2)) ## 결과값 : TtestResult(statistic=np.float64(2.2108140580092237), pvalue=np.float64(0.04914857789252186), df=np.float64(11.0))

## p-value(0.049) < 0.05 이므로 대립 채택 "반별 수학 평균 점수는 다르다"

## 번외) equal_val 옵션 주기 (True : 두 집단의 분산이 같다고 가정(기본) / False : 두 집단의 분산이 다르다고 가정)
print(stats.ttest_ind(class1, class2, equal_var=False))
## 결과값 : TtestResult(statistic=np.float64(2.1818699281825236), pvalue=np.float64(0.059589330071355334), df=np.float64(8.272682358753572))

## p-value(0.059) > 0.05 이므로 귀무 채택 "반별 수학 평균 점수는 같다"
## equal_var 옵션에 따라 채택 가설이 달라짐


## 문제 조건 : (단측 검정) 귀무 : 반별 수학 점수는 같다 / 대립 : 2반 수학 평균 점수가 더 높다 (모분산은 동일)
print(stats.ttest_ind(class1, class2, alternative='less', equal_var=True)) # 모분산 동일 equal_var=True / 2반이 더 높음 : class1 - class2 값 < 0 '더 작다' less
## 결과값 : TtestResult(statistic=np.float64(2.2108140580092237), pvalue=np.float64(0.9754257110537391), df=np.float64(11.0))

# p-value(0.975) > 0.05 이므로 귀무가설 채택 "반별 수학 점수는 같다", "2반 수학 평균 점수가 더 높다"는 통계적으로 유의미하지 않다

## 문제 조건 : (단측 검정) 귀무 : 반별 수학 점수는 같다 / 대립 : 1반 수학 평균 점수가 더 높다 (모분산은 동일)
print(stats.ttest_ind(class1, class2, equal_var=True, alternative='greater')) ## class1 - class2 > 0 "1반 더 높음"
## 결과값 : TtestResult(statistic=np.float64(2.2108140580092237), pvalue=np.float64(0.02457428894626093), df=np.float64(11.0))

# p-value(0.024) < 0.05 이므로 대립 채택 "1반 수학 평균 점수가 2반의 수학 평균 점수보다 더 높다"는 가설이 통계적으로 유의미하다
```

```python
import pandas as pd
class1 = [85, 90, 92, 88, 86, 89, 83, 87]
class2 = [80, 82, 88, 85, 84]
```

```python
from scipy import stats
print(stats.ttest_ind(class1, class2))  ## 독립 표본 검정 (서로 다른 두 집단) stats.ttest_ind(class1, class2) + equal_var(True/False), alternative=('greater'/'less)
```

```python
print(stats.ttest_ind(class1, class2, equal_var=False))
```

```python
print(stats.ttest_ind(class1, class2, equal_var=True, alternative='less'))
```

```python
print(stats.ttest_ind(class1, class2, equal_var=True, alternative='greater'))
```

### 2. 심화학습

```python
## 독립 표본 검정 수행 전 "등분산성 검증" - 두 집단 데이터에 정규성 검정 시행 (shapiro() 또 사용 후 lavene() 함수 사용 - 데이터셋 정규성 여부 평가)
## 통과 시 ttest_ind() 사용 / 통과 X : 비모수 검정 Mann-Whitney U 검정으로 분석 진행 / 두 집단 분산 다름 : Welch t-검정 실시
## 뭐야 이게

# 문제 : 귀무 "반별 수학 평균 점수는 같음" / 대립 "2반 수학 평균 점수가 더 높음"
import pandas as pd
class1 = [85, 90, 92, 88, 86, 89, 83, 87]
class2 = [80, 82, 88, 85, 84]

## 정규성 검정은 뭐다? stats.shapiro()
from scipy import stats
print(stats.shapiro(class1)) # 결과 : ShapiroResult(statistic=np.float64(0.9981893537736595), pvalue=np.float64(0.999986994137081)) "귀무 : 정규성 만족"
print(stats.shapiro(class2)) # 결과 : ShapiroResult(statistic=np.float64(0.9917398436295009), pvalue=np.float64(0.9854182266624983)) "귀무 : 정규성 만족"

## 등분산성 검정 levene()  - 두 집단 동시에 집어넣음
print(stats.levene(class1, class2)) # 결과 : LeveneResult(statistic=np.float64(0.0027925869510027727), pvalue=np.float64(0.958802951766629)) "귀무 : 등분산성 만족"
#### 독립 표본 : 정규 검정(shapiro : 각각) - 등분산성 검정(levene : 동시에) - 독립 표본 ttest(ttest_ind())

print(stats.ttest_ind(class1, class2, alternative='less')) # 결과 : TtestResult(statistic=np.float64(2.2108140580092237), pvalue=np.float64(0.9754257110537391), df=np.float64(11.0))
# p-value(0.975) > 0.05 이므로 귀무 채택 "반별 수학 평균 점수는 같다"는 가설이 통계적으로 유의미함
```

```python
import pandas as pd
class1 = [85, 90, 92, 88, 86, 89, 83, 87]
class2 = [80, 82, 88, 85, 84]

from scipy import stats
print(stats.shapiro(class1))
print(stats.shapiro(class2))
```

```python
print(stats.levene(class1, class2))
```

```python
print(stats.ttest_ind(class1, class2, alternative='less', equal_var=True))
```

```python
## 정규성 위반되는 데이터에 대한 독립 표본 검정 수행
import pandas as pd
class1 = [85, 90, 92, 88, 86, 89, 83, 87]
class2 = [80, 82, 88, 85, 130]

## 뭐다? 1. shapiro(각각) - 2. levene(동시에) : 등분산성 - 3. 독립 ttest_ind()
from scipy import stats
print(stats.shapiro(class1)) # 결과 ShapiroResult(statistic=np.float64(0.9981893537736595), pvalue=np.float64(0.999986994137081)) > 0.05 이므로 귀무 채택 (정규성 만족)
print(stats.shapiro(class2)) # 결과 ShapiroResult(statistic=np.float64(0.6880497349322277), pvalue=np.float64(0.007151570728885509)) <<< 0.05 이므로 대립 (정규성 불만족)

## 정규성 불만족 -> 맨휘트니 U 검정 (비모수 검정)
stats.mannwhitneyu(class1, class2, alternative='less') # 대립가설이 2반이 높다 이므로 class1 - class2 < 0 "less"
# 결과 : MannwhitneyuResult(statistic=np.float64(26.0), pvalue=np.float64(0.8299904236851448))

## p-value(0.829) > 0.05 이므로 귀무 채택 "반별 수학 평균 점수는 같다"는 가설이 통계적으로 유의미
## 어ㅜ 피곤해
```
