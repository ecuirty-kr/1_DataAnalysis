### Section1.

```python
## 단일 표본 검정 - 연습 문제
## 문제 정의 : 커피 제조회사의 새로 출시한 커피의 카페인 함량이 평균 95 mg 미만이라고 주장
## 사실인지 알아보기 위해 25 개의 커피 샘플을 무작위 추출. 주장이 타당한지 유의수준 5% 에서 검정
## 귀무가설 : u >= 95 mg / 대립가설 : u < 95 mg

# 문제 1. 표본 데이터의 평균
# 문제 2. shapiro-Wilk 검정의 p-value
# 문제 3. 단일 표본 t-검정의 검정 통계량
# 문제 4. 단일 표본 t-검정의 p-value
# 문제 5. 유의수준 0.05 하에서 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 입력

# 문제에서 제시한 데이터
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)': [
        94.2, 93.7, 95.5, 93.9, 94.0, 95.2, 94.7, 93.5, 92.8, 94.4,
        93.8, 94.6, 93.3, 95.1, 94.3, 94.9, 93.9, 94.8, 95.0, 94.2,
        93.7, 94.4, 95.1, 94.0, 93.6
    ]
})
df ## 데이터 프레임 확인

### 문제 요구사항 : 단일 표본 검정 - 평균 데이터 조건을 만족하는지 검정 (단일 표본 검정)
### t-검정이므로 ttest_1samp()
# 1. 표본 데이터 평균 구하기
print(df.mean()) # 결과 : Caffeine(mg)    94.264

# 2. shapiro-Wilk 검정의 p-value : shapiro()
## shapiro()는 정규성 검정 시 사용하는 함수 [라이브러리 scipy.stats]
from scipy import stats
print(stats.shapiro(df))
# 결과 : statistic=np.float64(0.9826578166170536), pvalue=np.float64(0.9322031137746971)
# 답안 : p-value = 0.9322031137746971

# 3. 단일 표본 t-검정의 검정통계량 [라이브러리 scipy.stats]
# t_statistic, p_value = stats.ttest_1samp(df)
## ** 문제 조건 숙지 : 커피 회사에서 카페인 함량이 95mg "미만"이라고 주장
print(stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less'))  ## 'less'에 옵션값 명 alternative= 왜 까먹음
# 결과 : statistic=np.float64(-5.501737036221897), pvalue=np.float64(5.8686553916715e-06       ## p-value 값이 지수형태로 출력
# 답안 : -5.501737036221897

# 4. 단일 표본 t-검정의 p-value
## 위에서 ttest_1samp() 한 결과에서 지수형태로 출력된 p-value 값을 실수값으로 출력
t_statistic, p_value = stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less')
print("{:.10f}".format(p_value))  # 결과 : 0.0000058687
# 답안 : 0.0000058687
## 실수 포맷을 출력하는 방법 숙지 : ("{:.10f}".format(값))

# 5. 유의수준 0.05 하에서 귀무가설을 기준으로 검정 결과를 기각/채택
# p-value(0.0000058687) < 0.05 이므로 귀무가설 기각
# 답안 : 기각
```

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)': [
        94.2, 93.7, 95.5, 93.9, 94.0, 95.2, 94.7, 93.5, 92.8, 94.4,
        93.8, 94.6, 93.3, 95.1, 94.3, 94.9, 93.9, 94.8, 95.0, 94.2,
        93.7, 94.4, 95.1, 94.0, 93.6
    ]
})
```

```python
# 1. 표본 평균
print(df.mean())

# 2. 정규성 검정
from scipy import stats
print(stats.shapiro(df['Caffeine(mg)']))

# 3~5.단일 표본 t-검정
print(stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less'))
```

```python
statistic, pvalue = stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less')
print("{:.10f}".format(pvalue))
```

### Section2.

```python
## 독립 표본 검정 - 연습문제
## 문제 정의 : 스마트폰 배터리의 충전 시간을 줄이기 위한 새로운 충전기 개발. 개발된 충전기 효과를 검정
## 사용자 집단을 두 그룹으로 나눠 한 그룹에는 새로운 충전기, 다른 그룹에는 기존의 충전기 제공. 두 그불의 평균 충전 완료 시간 비교
## 두 그룹의 충전 시간은 정규분포를 따르며 분산은 동일. 새 충전기가 더 빠르다고 할 수 있는지 유의수준 5% 하에서 검정
## u1 : 새로운 충전기 집단의 모평균 / u2 : 기존 충전기 집단의 모평균
## 귀무가설 : u1 >= u2 / 대립가설 : u1 < u2

## 문제 1. 위의 가설을 검정하기 위한 검정 통계량
## 문제 2. 위의 통계량에 대한 p-value
## 문제 3. 유의수준 0.05 하에서 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택

# 문제에서 제시한 데이터
import pandas as pd
df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
## 데이터 확인
print(df.head(5))

## 문제 요구사항 : 서로 다른 그룹 검정 - 평균값에 대한 비교 (1. 단일 - 평균을 만족하는지 / 2. 대응 - 같은 그룹에 대해 / 3. 독립 - 서로 다른 두 그룹)
## 독립 표본 검정 ttest_ind() 함수 / 라이브러리 scipy.stats
from scipy import stats

# 1. 검정 통계량 구하기 - ttest_ind() 사용하면 statistic 값도 구할 수 있음 : 얘가 검정통계량
## ttest_ind() 인자값 좀 외워라. 왜 또 까먹음
# 문제에서 기존 충전기의 충전시간과 새 충전기의 충전시간을 비교하라고 주어짐 -> 충전기 데이터 가 ['New'], ['Old']로 합쳐져 있으므로 분리해야 함
new_cond = df['충전기'] == 'New'
new_cond2 = df['충전기'] == 'Old'
print(stats.ttest_ind(df[new_cond]['충전시간'], df[new_cond2]['충전시간'],alternative='greater')) ## ttest_ind(class1, class2, alternative) 1-2 값이 클 지 작을 지 : 새 충전기 빠르면 1-2 > 0
## *** df[cond]['충전시간'] -> cond 조건 넣어주고 비교할 컬럼명 '충전시간' 까지 지정해줘야 함 ***
# 결과 : statistic=np.float64(-4.582575694955849), pvalue=np.float64(0.999884534522123), df=np.float64(18.0)
# 답안 : -4.582575694955849

# 2. 위의 통계량에 대한 p-value
t_statistic, p_value = stats.ttest_ind(df[new_cond]['충전시간'], df[new_cond2]['충전시간'], alternative='greater', equal_var=True) # 모분산 동일 : equal_var=True (기본값 : 생략가능)
print("p-value : ", p_value)
# 결과 : p-value :  0.999884534522123

# 3. 유의수준 0.05 하에서 귀무가설을 기준으로 검정 결과를 기각/채택 선택
## p-value(0.99988) > 0.05 이므로 귀무가설 채택
# 답안 : 채택

### ** 2, 3 번 오답 : 개발된 충전기 성능이 좋으려면 값이 "작아야" 함 -> class(new) - class(old) 값 < 0 이 되어야 하므로 alternative='less'가 맞음

## 다시ㅜ 풀기
# 1. 검정 통계량, 2. p-value 구하기 (독립 표본 검정)
print(stats.ttest_ind(df[new_cond]['충전시간'], df[new_cond2]['충전시간'], alternative='less'))
# 결과 : TtestResult(statistic=np.float64(-4.582575694955849), pvalue=np.float64(0.00011546547787696304), df=np.float64(18.0))
# 1 답안 : -4.582575694955849
# 2 답안 : 0.00011546547787696304

# 3. 채택 / 기각
# p-value(0.00011) < 0.05 이므로 귀무가설 기각
# 답안 : 기각
```

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
print(df.head(2))
```

```python
# 1~3. 독립 표본 t-검정
new_cond = df['충전기'] == 'New'
old_cond = df['충전기'] == 'Old'
print(stats.ttest_ind(df[new_cond]['충전시간'], df[old_cond]['충전시간'], alternative='less', equal_var=True))
```

```python
## 대응 표본 검정 - 연습 문제 (같은 집단에 대해 전후 시간차 데이터 비교)
## 문제 조건 : 새로 개발한 교육과정에 참여하면 학습 시간이 단축된다고 주장. 10명이 학생들을 대상으로
## 기본 교육 방법과 새로운 교육 방법을 적용한 결과에 대해 유의수준 5% 에서 새로운 교육 방법이 교육 시간을 단축시켰는지 검정
## 귀무가설 : u = 0 / 대립가설 : u < 0 (u0 = (새로운 교수 방법 - 기존 교수 방법)의 평균)

## 문제 1. u0 의 표본 평균을 구하시오
## 문제 2. 위의 가설을 검정하기 위한 검정 통계량
## 문제 3. 위의 통계량에 대한 p-value
## 문제 4. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택

# 문제에서 제시한 데이터
import pandas as pd
df = pd.DataFrame({
     'User': list(range(1, 11)),
    '기존방법': [60.4, 60.7, 60.5, 60.3, 60.8, 60.6, 60.2, 60.5, 60.7, 60.4],
    '새로운방법': [59.8, 60.2, 60.1, 59.9, 59.7, 58.4, 57.0, 60.3, 59.6, 59.8]
})
## 데이터 확인
print(df.head(5))

# 1. u0의 표본 평균 - 문제에서 u0 = (새로운 교수 방법 - 기존 교수 방법)의 평균 이라고 제시함
# cond = (df['기존방법'] - df['새로운방법']) # 아님. 새로운 데이터 프레임 생성
df['diff'] = df['기존방법'] - df['새로운방법']
print(df['diff'].mean()) # 결과 : 1.0300000000000005
# 답안 : 1.0300000000000005

# 2. 위의 가설을 검정하기 위한 검정 통계량 (대응 표본 검정이므로 ttest_rel())
from scipy import stats
print(stats.ttest_rel(df['새로운방법'], df['기존방법'], alternative='less')) ## 함수 인자값 좀 외워라 ttest_rel(class1, class2, alternative)
## 학습 시간 단축에 관한 질문이므로 c1 - c2 값 < 0 이어야 함 'less' + 문제에서 제시 "대립가설 u < 0" = "작아야 한다"
# 결과 : statistic=np.float64(-3.407973078114844), pvalue=np.float64(0.0038872633380070652), df=np.int64(9)
# 답안 : -3.407973078114844

# 3. 위의 통계량에 대한 p-value
t_statistic, p_value = stats.ttest_rel(df['새로운방법'],df['기존방법'], alternative='less')
print(p_value) # 결과 : 0.0038872633380070652
# 답안 : 0.0038872633380070652

# 4. 유의수준 0.05 하에서 귀무가설 기준으로 검정결과 채택/기각
# p-value(0.0038) < 0.05 이므로 귀무가설 기각
# 답안 : 기각
```

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    'User': list(range(1, 11)),
    '기존방법': [60.4, 60.7, 60.5, 60.3, 60.8, 60.6, 60.2, 60.5, 60.7, 60.4],
    '새로운방법': [59.8, 60.2, 60.1, 59.9, 59.7, 58.4, 57.0, 60.3, 59.6, 59.8]
})
print(df.head(2))
```

```python
# 1. 표본 평균
df['diff'] = df['새로운방법'] - df['기존방법']
print(df['diff'].mean())

# 2~4. 대응 표본 t-검정
print(stats.ttest_rel(df['새로운방법'], df['기존방법'], alternative='less'))
```

```python
## 적합도 검정 - 연습 문제 (카이제곱 적합도 검정 : 선형성을 만족하는지 검정) 사용 함수 : chisquare()
## 문제 조건 : 한 도시의 운전자 1,000 명을 대상으로 교통사고 경험 수 조사
## 1회 : 550명 / 2회 : 250명 / 3회 : 100명 / 4회 : 70명 / 5회 이상 : 30명
## 전국 데이터 기반 교통사고 경험 수 분포 1회 : 60% / 2회 : 25% / 3회 : 8% / 4회 : 5% / 5회 이상 : 2%
## 이 도시의 데이터가 전국적인 경향을 따르는지 검정 (유의수준 0.05)

## 문제 1. 이 도시의 교통사고 경험자 중 5회 이상의 비율을 0과 1사이로 구하시오
## 문제 2. 이 도시 운전자들의 교통사고 경험 수 분포가 전국적인 경향을 따르는지 검정하기 위한 검정 통계량
## 문제 3. 위의 통계량 대한 p-value
## 문제 4. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택

import pandas as pd
# 문제에서 주어진 데이터
## 구분 | 경험 수 | 전국적인 경험(%)
## 1회  |  550    | 60%
## 2회  | 250     | 25%
## 3회  | 100     | 8 %
## 4회  | 70      | 5%
## 5회  | 30      | 2%

# 1. 교통사고 경험자 중 5회 이상의 비율 (0~1 사이로 계산)
print(30/1000)   ## 어렵게 생각하지 말고 '해당 조건 인수/전체 인수'
# 결과 : 0.03
# 답안 : 0.03

# 2, 3, 4. 검정 통계량, p-value, 기각/채택 - 카이제곱 적합도(조사, 평균 데이터 비율 적합한지 검정) [함수 : chisquare()]
from scipy.stats import chisquare ## from에 scipy.stats 묶어서 사용 import 함수
## 문제에서 주어진 조건의 단위가 다르므로 맞춰줘야 함(빈도로 맞춤. 비율(X)->빈도 변환)
observed = [550, 250, 100, 70, 30]
expected = [1000*0.6, 1000*0.25, 1000*0.08, 1000*0.05, 1000*0.02]  ## 비율 -> 빈도 변환은 "전체 조사 인구수*해당 비율" 로 계산 ***** 외우라곸!!
print(chisquare(observed, expected))
# 결과 : statistic=np.float64(22.166666666666668), pvalue=np.float64(0.00018567620386641427)
# 2 답안 : 22.166666666666668
# 3 답안 : 0.00018567620386641427

# p-value(0.0001) < 0.05 이므로 귀무가설 기각(대립가설 채택)
# 4 답안 : 기각
```

```python
# 1. 교통사고 5회 이상 경험 비율
print(30 / 1000)

# 2~4. 적합도 검정
from scipy.stats import chisquare
observed = [550, 250, 100, 70, 30]
expected = [1000*0.60, 1000*0.25, 1000*0.08, 1000*0.05, 1000*0.02]
print(chisquare(observed, expected))
```

```python
### 카이제곱 독립성 검정 (두 그룹이 서로 독립적인지 연관성을 갖는지 검정) 사용 함수 : chi2_contingency() 아 인자값
## 문제 조건: 빅데이터 분석기사 캠프와 정보처리기사 캠프 존재. 최근 두 캠프 학생 사이에서 주최한 세미나에 참여하기 위한 등록 필요
## 각 캠프의 세미나 등록 여부를 조사한 결과를 갖고 캠프와 세미나 등록 여부가 서로 독립적인지 검정 (유의수준 0.05)
## 귀무 : 캠프와 세미나 등록 여부는 독립적 / 대립 : 캠프와 세미나 등록 여부는 서로 독립적이지 않다

## 교차표 데이터가 주어짐
# 문제 1. 캠프와 세미나 등록 여부가 서로 독립적인지 검정하기 위한 검정 통계량
# 문제 2. 위 통계량에 대한 p-value
# 문제 3. 검정 결과를 유의수준 0.05 하에서 귀무가설 기준으로 채택/기각

# 주어진 데이터
## 구분 : 빅데이터 분석기사 캠프 '등록' 50  / '등록 안함' 30
## 구분 : 정보처리기사 캠프      '등록' 60  / '등록 안함' 40

# 1. 캠프와 세미나 등록 여부가 서로 독립적인지 검정하기 위한 검정 통계량
import pandas as pd
from scipy.stats import chi2_contingency

## 카이제곱 독립성 검정 전 교차표 데이터 프레임 생성
# observed = pd.DataFrame({[50, 30], [60, 40]})
observed = pd.DataFrame([[50, 30], [60, 40]])
print(chi2_contingency(observed)) ## 데이터 프레임 생성을 잘못함 -> 중괄호 아니고 대괄호 두 번임

# 결과 : statistic=np.float64(0.03535714285714309), pvalue=np.float64(0.8508492527705047), dof=1, expected_freq=array([[48.88888889, 31.11111111], [61.11111111, 38.88888889]]
# 답안 : 0.03535714285714309

# 2. p-value 답안 : 0.8508492527705047
# 3. p-value(0.8508) > 0.05 이므로 귀무가설 채택 / 답안 : 채택
```

```python
#### 위와 같은 문제, 로우(row) 데이터가 주어졌을 때
import pandas as pd
df = pd.DataFrame({
    '캠프' : ['빅분기']*80 + ['정처기']*100,  ## 문제에서 주어진 각 캠프의 '등록', '등록여부' 값을 모두 합해서 캠프에 곱하는거임.... Aㅏ
    '등록여부' : ['등록']*50 + ['등록안함']*30 + ['등록']*60 + ['등록안함']*40
})
## 데이터 확인
print(df.head(5))

## 독립성 검정 수행 전 로우 데이터를 교차표 데이터로 전환
df = pd.crosstab(df['캠프'], df['등록여부'])  ## 교차표에 사용할 데이터 프레임 명을 직접 넣어주는 것임!

#  카이제곱 독립성 검정 수행 chi2_contingency()
from scipy.stats import chi2_contingency
print(chi2_contingency(df))
# 결과 : statistic=np.float64(0.03535714285714309), pvalue=np.float64(0.8508492527705047), dof=1, expected_freq=array([[48.88888889, 31.11111111], [61.11111111, 38.88888889]])
# 1 답안 : 0.03535714285714309
# 2 답안 : 0.8508492527705047 (p-value > 0.05)
# 3 답안 : 귀무가설 채택
```

```python
### 선형 회귀 - 연습 문제 (독립변수가 두 개 이상인 다중 선형 회귀)
### 문제 조건 : 한 도시의 배달 서비스 회사에서 한 달 동안의 주문 데이터를 수집. 배달 주문량은 할인율(%), 온도(C), 광고비(천 원)
### 다양한 요인들에 영향을 받음. 배달 서비스 회사는 이런 요인을 고려해 주문량을 예측하고자 함.
### 할인율, 온도, 광고비를 독립변수로 사용하여 다중 선형 회귀 모델을 구축하고, 문제에 답하시오

## 문제 1. 할인율과 온도의 상관계수 (반올림하여 소수 둘째자리까지)
## 문제 2. 모델의 결정 계수 (반올림하여 소수 둘째 자리까지)
## 문제 3. 각 변수의 회귀 계수 (반올림하여 소수 넷째 자리까지)
## 문제 4. 모델의 절편(Intercept) (반올림하여 소수 넷째 자리까지)
## 문제 5. 온도의 회귀 계수가 통계적으로 유의한지 검정 (p-value를 반올림하여 소수 넷째자리까지)
## 문제 6. 데이터가 할인율이 10%, 온도가 20C, 광고비가 500,000 원일 때 배달 주문량 예측
## 문제 7. 잔차 제곱합
## 문제 8. MSE
## 문제 9. 온도의 회귀 계수에 대한 90% 신뢰구간
## 문제 10. 데이터가 할인율이 15%, 온도가 25C, 광고비가 300,000 일 때, 신뢰구간과 예측구간
## 문제 11. 독립변수 할인율과 온도를 고정한 상태에서 광고비가 배달 주문량에 영향을 주는지 가설 검정 수행. 유의수준 0.05 기준으로 기각/채택
## 귀무 : 광고비의 회귀 계수는 0 (광고비가 배달 주문량에 영향을 주지 않는다) / 대립 : 광고비 회귀 계수는 0이 아니다 (광고비가 배달 주문량에 영향)

# 문제에서 주어진 데이터
import pandas as pd
df = pd.DataFrame({  ## 데이터는 복붙
    '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25,
            4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 2],
    '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10,
           29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
    '광고비': [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642,
            591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
    '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893,
            830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
## 데이터 확인
print(df.head(5))

## 로지스틱 회귀 함수 사용 : scipy.stats.logit() -> 아님
# from statsmodels.formula.api import ols
from statsmodels.formula.api import ols

# 1. 할인율과 온도의 상관계수
model = ols('주문량~할인율+온도+광고비', data=df).fit()
# 상관 계수 : 데이터프레임.corr(데이터 프레임) - 왜!! 어제 한 걸!!!!! 기억을 모태!!!!!
print(round(df['할인율'].corr(df['온도']),2)) ## round(값, 자리수) 함수에 몇 째 자리인지 인수 빼먹기
# 결과 : 0.09 / 답안 : 0.09

# 2. 모델의 결정계수 rsquare() - d 좀 빼먹지좀 마라고
print(round(model.rsquared,2)) ## 결과 : 0.4
# 답안 : 0.4

# 3. 각 변수의 회귀 계수
print(round(model.params, 4))
# 결과 + 답안 : 할인율  4.2068 / 온도  9.4798 / 광고비   0.4148

# 4. 모델의 절편(Intercept)
print(round(model.params,4)) # 결과 중 Intercept = Intercept  267.6609
# 답안 : 267.6609

# 5. 온도의 회귀 계수가 통계적으로 유의한지 검정 (p-value를 반올림하여 소수 넷째자리까지)
print(round(model.pvalues['온도'], 4))  ## pvalues 사용법 좀. model.pvalues['온도']
# 결과 및 답안 : 0.0289

# 6. 데이터가 할인율이 10%, 온도가 25C, 광고비 500,000 일 때 배달 주문량 예측
# 새 데이터 프레임 생성 (데이터->데이터 프레임 (X) / 데이터 프레임 직접 생성)
new_data = pd.DataFrame({
    '할인율' : [10],
    '온도' : [20],
    '광고비' : [500]  ## 문제에서 500,000라고 주어졌다고 그대로 넣지말고 단위 확인하기 - 값 크게 달라짐
})
# new_model = ols('주문량~할인율+온도+광고비', data=new_df).fit() ## 이게 아님. 뻘짓을...
result = model.predict(new_data)
print(int(result[0])) ## 인덱스 번호까지 출력되므로 [0]로 값만 출력 + 실수값이고, 반올림 조건 없으므로 int() 에 넣어서 정수로 출력
# 답안 : 706

# 7. 잔차 제곱 합
# 잔차 먼저 구하고 잔차 제곱 값을 sum() : 잔차 = 실제값 - 예측값 [은 맞는데 df[종속]-model.predict(df) 도 같이 외우기]
df['잔차'] = df['주문량'] - model.predict(df)
print(sum(df['잔차']**2)) # 결과 : 732197.8988142072
# 답안 : 732197.8988142072

# 8. MSE (평균 제곱 오차) 여기서도 잔차 사용 : 잔차 제곱의 평균 구하는 것이다
df['잔차'] = df['주문량'] - model.predict(df)
print((df['잔차']**2).mean()) # 결과 : 24406.596627140243
# 답안 : 24406.596627140243

# 9. 온도의 회귀 계수에 대한 90% 신뢰구간 - 얘는 왜 기억이 안 나냐
## model.conf_int() 함수 사용 : 신뢰구간(%) 이므로 alpha 옵션값을 갖는 함수임 [90% = alpha=0.1 (10%를 의미)]
print(model.conf_int(alpha=0.1)) # 모든 회귀 계수에 대한 90% 신뢰구간이 출력됨
# 결과 중 온도 값 : 온도    2.490702   16.468984
# 답안 : 2.490702   16.468984
## 기억 좀 해라 conf_inf(alpha=값) : 신뢰구간 계산

# 10. 데이터가 할인율이 15%, 온도가 25C, 광고비가 300,000 일 때, 90% 신뢰 구간과 예측 구간
# predict_frame(), get_머 있었는데 뭐였지 -> 다 틀림
## 새 데이터 주어졌으므로 새 데이터 프레임 생성
new_data = pd.DataFrame({
    '할인율' : [15],
    '온도' : [25],
    '광고비' : [300]
})
pred = model.get_prediction(new_data) ## 출력하는게 아니라 get_prediction(dataFrame) 수행하고 예측 값을 pred 변수에 담아야 함
result = pred.summary_frame(alpha=0.1) ## summary_frame(alpha) 함수는 인자값으로 alpha=값 주는거임. 신뢰구간 구하는거잖슴. model이 아니라 pred 만든거에 pred.summary_frame() 쓰는거임
print(result) # 테이블 형태로 출력됨
## 결과에서 확인 : (신뢰구간) mean_ci_lower  mean_ci_upper  / (예측 구간) obs_ci_lower obs_ci_upper
# 답안 : 신뢰구간 = 598.566925   785.847846  / 예측 구간 : 334.777165   1049.637606

# 11. 독립변수 할인율과 온도를 고정한 상태에서 광고비가 배달 주문량에 영향을 주는지 가설 검정 수행. 유의수준 0.05 기준으로 기각/채택
# 고정값 안 하고 넘어갔었는데 이걸 내네.
cond = model.pvalues['광고비'] < 0.05
if cond:
  result = "기각"
else:
  result = "채택"
print(result) # 결과 : 기각
# 답안 : 기각

## 10 번 문제 뭘 잘못 했길래 답이 다른거냐.
## 그거슨 90% 신뢰구간을 요구했는데 유의수준 0.05 라고 alpha=0.05로 해버렸기 때문. (고침)
# 완전 오래 걸리네
```

```python
# 다중 선형 회귀 모델 적합
from statsmodels.formula.api import ols
model = ols('주문량 ~ 할인율 + 온도 + 광고비', data=df).fit()

# 1. 상관계수
print("1. 상관계수:", round(df['할인율'].corr(df['온도']), 2))

# 2. 결정계수
print("2. 결정계수(R-squared):", round(model.rsquared, 2))

# 3. 회귀계수(기울기)
print("3. 회귀계수:", round(model.params, 4))

# 4. 절편
print("4. 절편:", round(model.params['Intercept'], 4))

# 5. 회귀계수 검정
print("5. pvalue:", round(model.pvalues['온도'], 4))

# 6. 예측 판매량
new_data = pd.DataFrame({"할인율": [10], "온도": [20], "광고비": [500]})
result = model.predict(new_data)
print("6. 새로운 데이터:", int(result[0]))

# 7. 잔차 제곱합
df['잔차'] = df['주문량'] - model.predict(df)
print("7. 잔차 제곱합:", round(sum(df['잔차']**2), 2))

# 8. MSE(Mean Squared Error)
MSE = (df['잔차'] ** 2).mean()
print('8. MSE:', round(MSE, 4))

# 9. 각 변수에 대한 90% 신뢰구간
print("9. 신뢰구간:\n", model.conf_int(alpha=0.1))

# 10. 새로운 데이터의 예측값의 90% 신뢰구간과 예측구간
new_data = pd.DataFrame({"할인율": [15], "온도": [25], "광고비": [300]})
pred = model.get_prediction(new_data)
result = pred.summary_frame(alpha=0.1)
print("10. 예측값의 신뢰구간과 예측구간:\n", result)

# 11. 광고비는 배달 주문량에 영향을 주는지 가설 검정
cond = model.pvalues['광고비'] < 0.05
if cond:
    result = "기각"
else:
    result = "채택"
print("11. 귀무가설", result)

# 선형 회귀 모델의 요약 결과
print(model.summary())
```

```python
import pandas as pd
# df = pd.read_csv("customer_travel.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/refs/heads/main/part3/ch6/customer_travel.csv")

df.head()
```

```python
# 데이터 분할
midpoint = len(df) // 2 #몫(정수)
a = df.iloc[:midpoint]
b = df.iloc[midpoint:]

# 데이터 확인
a.shape, b.shape
```

```python
# 1) 유의하지 않은 독립변수의 개수
from statsmodels.formula.api import logit
formula = "target ~ age + service + social + booked"
model = logit(formula, data=a).fit()
# print(model.summary())
print("1.", sum(model.pvalues[1:] >= 0.05)) # 상수항(Intercept) 제외

# 2. 수정된 모델에서 가장 큰 p-value를 가진 변수의 이름
formula = 'target ~ age + booked'
model = logit(formula, data=a).fit()
# print(model.summary())
print("2.", model.pvalues[1:].idxmax()) # 상수항(Intercept) 제외

# 3. 수정된 모델에서 독립변수 중 절댓값이 가장 큰 회귀계수를 가진 변수의 이름
print("3.", model.params[1:].abs().idxmax())

# 4. 로그 우도
print("4.", model.llf)

# 5. 잔차이탈도
print("5.", -2 * model.llf)

# 6. 'booked' 변수가 3 증가할 때 오즈비
import numpy as np
print("6.", np.exp(model.params['booked'] * 3))

# 7. p-value가 0.05보다 작은 회귀계수의 총합
print("7.", model.params[model.pvalues < 0.05].sum()) # 상수항(절편)도 포함

# 8. 정확도
pred = model.predict(b)
pred = (pred > 0.5).astype(int)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(b['target'], pred)
print("8.", accuracy)

# 9. 오류율
error_rate = 1 - accuracy
error_rate
print("9.", error_rate)
```
