```python
#### 가능?
#### 문제 조건: (별도 표기 없음)

# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/loan.csv")
print(df.head(4)) ## 데이터 확인

## 문제 1. 총 대출액은 신용대출과 담보대출의 합이다.
# 새 컬럼 생성
df['총대출액'] = df['신용대출'] + df['담보대출']
print(df.head()) ## 왜 총대출액 공란임

## 문제 2. 각 지역코드 내에서, 성별(예: 1과 2)별로 '총대출액'의 합계를 계산하시오
# cond1 = df['성별'] == 1
# cond2 = df['성별'] == 2
## 지역코드와 성별"별"로 총대출액 합계 계산 : groupby로 2 가지 그룹화 수행
grouped = df.groupby(['지역코드', '성별'])['총대출액'].sum().unstack() ## 지역코드, 성별 모두 groupby로 그룹화 / 한 데이터의 ['총대출액']의 / 합계 .sum() / unstack() : 열 형태로 변환

## 문제 3. 성별 간 총대출액의 차이가 가장 큰 지역코드
# 위에서 만든 grouped 사용
grouped['차이'] = abs(grouped[1] - grouped[2])  ## 성별 간 총대출액 차이를 계산하여 '차이' 컬럼으로 생성
result = grouped['차이'].idxmax()
print(result) # 결과 : 4100000278
# 답안 : 4100000278

## 고난도 회차라고 명시되어 있음
```

```python
#### 문제 조건 : (별도 표기 없음)
## 문제 1. 검거율은 검거건수 / 발생건수 이다.
# 새 컬럼 생성 문제
# 1. 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")
print(df.head(5)) # 데이터 확인

# 검거율 컬럼 생성
cond1 = df['구분'] == '검거건수'
cond2 = df['구분'] == '발생건수'
# df['검거율'] = (df.loc(cond1, '구분') / df.loc(cond2, '구분')) ## ...?

# 각각 발생건수와 검거건수만 가져오기 -> 그리고 계산 (나누어서 단계적 수행)
df1 = df[cond1].iloc[:, 2:] # 발생건수
df2 = df[cond2].iloc[:, 2:] # 검거건수

# 검거율 계산
df1 = df1.reset_index(drop=True) ## .reset_index(drop=true) : NaN 값 발생 시 인덱스 초기화 수행 (연산이 맞지 않는 경우 대비)
df2 = df2.reset_index(drop=True)
df3 = df2 / df1
print(df3)

## 2. 각 연도에서 가장 높은 검거율을 기록한 범죄유형을 찾으시오
# 각 연도 df['연도'] / 검거율 : df3 / 가장 높은 .max() OR .idxmax() / 범죄유형 -> idxmax로 행 가져와서 탐색
shape = df3.idxmax(axis=1) ## ['연도'] 대신 열 axis=1 값으로 지정
print("\n\n",shape) ### 놀랍게도. 봐도 잘 모르겠다는거.
# 0. 폭력범죄 가 가장 높은 검거율을 기록한 듯 !

## 3. 해당 범죄유형의 검거건수를 구한 후, 이 값들을 모두 합산하시오 (정수)
result = 0
for index, item in enumerate(shape): ## 위에서 구한 범죄유형 값으로 반복문 수행
  result = result + df2.loc[index, item] ## df2.loc[index, item] : 검거건수
print(result) # 5250
# 답안 : 5250 - 답 보고 했는데 왜 답이 다르냐요@@@@
```

```python
#### 문제 3
## 문제 3-1. 만족도의 결측치는 만족도의 평균값으로 대체
## 문제 3-2. 근속연수의 결측치는 같은 부서 내에서 동일한 성과등급을 가진 직원들의 평균 근속연수로 대체. 이때 평균 근속연수는 소수점 이하 절삭, 정수로
## 문제 3-3. 각 직원의 '연봉 / 근속연수' 값을 계산하여, 그 값이 세 번째로 큰 직원의 근속연수 ?
## 문제 3-4. 각 직원의 '연봉 / 만족도' 값을 계산하여, 그 값이 두 번째로 큰 직원의 교육 참가 횟수
## 문제 3-5. 위 3, 4번 값을 더한 결과 ? (정수)

# 라이브러리 및 데이터
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")
print(df.head(4)) # 데이터 확인

# 1. 만족도 결측치 확인
print(df['만족도'].isnull().sum()) # 120 개
avg = df['만족도'].mean()
print(avg) # 평균값
# df['만족도'].fillna(avg) # 평균값으로 대체
df['만족도'] = df['만족도'].fillna(avg)
print(df['만족도'].isnull().sum()) # 재확인 // ? 왜 대체 안 됨. -> .fillna() 수행하고 df['만족도'] = 까지 해줘야 적용이 됨

# 2. 근속연수의 결측치는 같은 부서 내에서 동일한 성과등급을 가진 직원들의 평균 근속연수로 대체
# 같은 부서 내 groupby('부서') / 동일한 성과등급 - 얘도 groupby 인가 / 평균 근속연수 df['근속연수'].mean()
average = df.groupby(['부서', '성과등급'])['근속연수'].transform('mean') ##  같은 부서 & 성과 / 의 근속연수 / 의 평균
print(df['근속연수'].isnull().sum()) # 결측치 확인부터   ** groupby 컬럼 두 개 묶을 때 대괄호[] 로 한 번 묶어야 함
# 결측치 : 48

# 평균 근속연수 -> 정수형
print(average.astype(int)) # 정수 변환 . astype(int) 하고 결측치 대체 fillna
aver = average.astype(int) # 변수에다 저장
df['근속연수'] = df['근속연수'].fillna(aver) ## error : type dismatch. 아.  ** average 변수 생성 시 평균값을 transform('mean') 으로 해줘야 타입 변환됨
print(df['근속연수'].isnull().sum()) ## 결측치 제거됨

############ 문제 너무 잘 냈다. 어렵다.

# 3. 각 직원의 '연봉 / 근속연수' 값을 계산하여 값이 세 번째로 큰 직원의 근속연수 구하기
## 연봉 (정수) / 근속연수 (실수) -> 둘 다 수치형이라 연산에는 문제 X // 값이 큰 순서대로~ 문제 : .nlargest(숫자)
# df['값'] = df['연봉'] / df['근속연수']
# idx = df['값'].nlargest(4)
# print(df['값']) ## 왜 값이 지수 형태로...
# print(idx)

df['연봉_근속연수'] = df['연봉'] / df['근속연수'] ## 수식 자체는 할만한데
df_year = df.nlargest(3, '연봉_근속연수') ## 활용이 문제임. '연봉_근속연수' 컬럼 값이 3 번째로 큰 데이터를 df_year에 저장
A = df_year.iloc[-1]['근속연수'] ## 이게 뭔데.  (iloc[-1] : 마지막 행 - 역행 개념으로 -1 임. 마지막 행이 세 번째 데이터)
print(A) # 결과 : 1.0
# 답안 : 1.0
###... 아니 교재에도 왜 답이 5 번만 나와있는건데.! 1.0 맞냐.

# 4. 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육 참가 횟수
df['연봉_만족도'] = df['연봉'] / df['만족도'] ## 연봉 / 만족도
df_like = df.nlargest(2, '연봉_만족도') # 연봉_만족도 계산한 값이 두 번째로 높은 데이터까지(1, 2 순위 출력)
B = df_like.iloc[-1]['교육참가횟수'] ## 이 사람의 교육참가횟수 (이때 행으로 지정 : 위에서 만든 변수 사용)

# 5. 문제 3 답 + 문제 4 답
print(A + B) # 7.0
# 답안 : 7.0

### .... 이게 맞나. 1 유형 시간 엄청 드는게. 복습 한 번 돌려야디
```

```python
# 방법2
import pandas as pd
df = pd.read_csv("crime.csv")
# df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")

# 1. 데이터를 길게 변환
df_long = pd.melt(df, id_vars=["연도", "구분"], var_name="범죄유형", value_name="건수")

# 2. 피벗(pivot)하여 발생건수와 검거건수를 한 행에 모으기:
df_pivot = df_long.pivot_table(index=["연도", "범죄유형"], columns="구분", values="건수").reset_index()

# 3. 검거율 계산
df_pivot["검거율"] = df_pivot["검거건수"] / df_pivot["발생건수"]

# 4. 각 연도별로 검거율이 가장 높은 범죄유형 찾기
listbox = df_pivot.groupby("연도")["검거율"].idxmax()
max_crime = df_pivot.loc[listbox]

# 5. 각 연도별 최고 검거율 범죄유형의 검거건수를 모두 합산하기
result = max_crime["검거건수"].sum()
result
```

```python
import pandas as pd
df = pd.read_csv("hr.csv")
# df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")

# 1) 만족도 결측치 처리
m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

# 2) 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움)
gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
gm = gm.astype(int)
df["근속연수"] = df["근속연수"].fillna(gm)

# 3) 연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
df_year = df.nlargest(3,'연봉_근속연수')
A = df_year.iloc[-1]["근속연수"]

# 4) 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.nlargest(2,'연봉_만족도')
B = df_like.iloc[-1]["교육참가횟수"]

# 5) 결과 출력
result = A + B
result
```

```python
import pandas as pd
df = pd.read_csv("hr.csv")
# df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")

# 1. 만족도 결측치 처리
m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

# 2. 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움)
gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
gm = gm.astype(int)
df["근속연수"] = df["근속연수"].fillna(gm)

# 3. 연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
df_year = df.sort_values(by="연봉_근속연수", ascending=False)
A = df_year.iloc[2]["근속연수"]

# 4. 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.sort_values(by="연봉_만족도", ascending=False)
B = df_like.iloc[1]["교육참가횟수"]

# 5. 결과 출력
result = A + B
result
```
