## Section4. 자료형 변환

```python
# 데이터 준비
data = {
    "메뉴":['아메리카노', '카페라떼', '카페모카', '카푸치노', '에스프레소', '밀크티', '녹차'],
    "가격":[4500.0, 5000.0, 5500.0, 5000.0, 4000.0, 5900.0, 5300.0],
    "칼로리":['10', '110', '250', '110', '20', '210', '0'],
}
df = pd.DataFrame(data)
df.info()
```

```python
### 데이터 생성
data = {
    "MENU":['아메리카노','카페라떼','카페모카','카푸치노','에스프레소','밀크티','녹차'],
    "PRICE":[4500.0, 5000.0, 5500.0, 5000.0, 4000.0, 5900.0, 5300.0],
    "KCAL":['10','110','250','110','20','210','0']
}
df = pd.DataFrame(data)
df.info() ## Non-Null Count 개수가 3개 모두 동일한 지 확인 + 인덱스 번호랑 column이름 정확한지 확인, 자료형 확인
```

```python
# 자료형 변환 / astype /  object -> int
df['가격'] = df['가격'].astype('int')
df.info()
```

```python
df['칼로리'] = df['칼로리'].astype('float')
df.info()
```

## Section20.피벗테이블

```python
import pandas as pd

# 직원별 부서 및 급여 정보
data = {
    '이름': ['서아', '다인', '채아', '예담', '종현', '태헌'],
    '부서': ['개발', '기획', '개발', '기획', '개발', '기획'],
    '급여': [3000, 3200, 3100, 3300, 2900, 3100]
}
df = pd.DataFrame(data)
print("[원본 데이터]")
print(df)

# 부서별 평균 급여 계산
pt = df.pivot_table(index='부서', values='급여', aggfunc='mean')
print("\n[부서별 평균 급여]") #\n은 줄 바꿈(새로운 줄 시작)
print(pt)
```

```python
# 부서와 직급별 급여 정보
data = {
    '부서': ['개발', '기획', '기획', '기획', '개발', '개발'],
    '직급': ['대리', '과장', '대리', '과장', '대리', '과장'],
    '급여': [3000, 4000, 3200, 4200, 3500, 4500]
}
df = pd.DataFrame(data)
print("[원본 데이터]")
print(df)

# 부서 및 직급별 급여 합계 계산
pt = df.pivot_table(index='부서', columns='직급', values='급여', aggfunc='sum')
print("\n[부서 및 직급별 급여 합계]")
print(pt)
```

```python
# 부서와 성별, 근속연수 정보
data = {
    '부서': ['개발', '기획', '기획', '기획', '개발', '개발'],
    '성별': ['남', '여', '남', '여', '남', '여'],
    '근속연수': [3, 5, 4, 6, 7, 8]
}
df = pd.DataFrame(data)
print("[원본 데이터]")
print(df)

# 부서 및 성별 평균 근속연수 계산
pt = df.pivot_table(index='부서', columns='성별', values='근속연수', aggfunc='mean')
print("\n[부서 및 성별 평균 근속연수]")
print(pt)
```

```python
# 다양한 집계를 위한 데이터셋
import pandas as pd

df = pd.DataFrame({
    "구분": ["전자", "전자", "전자", "전자", "전자", "가전", "가전", "가전", "가전"],
    "유형": ["일반", "일반", "일반", "특수", "특수", "일반", "일반", "특수", "특수"],
    "크기": ["소형", "대형", "대형", "소형", "소형", "대형", "소형", "소형", "대형"],
    "수량": [1, 2, 2, 3, 3, 4, 5, 6, 7],
    "금액": [2, 4, 5, 5, 6, 6, 8, 9, 9]
})
print(df)
```

```python
# 단일 값에 대해 집계
pt = pd.pivot_table(
    df,
    values='수량',
    index=['구분', '유형'],
    columns=['크기'],
    aggfunc="sum"
)
pt
```

```python
# 단일 값에 대해 집계 (누락 값 0 처리)
pt = pd.pivot_table(
    df,
    values='수량',
    index=['구분', '유형'],
    columns=['크기'],
    aggfunc="sum",
    fill_value=0
)
pt
```

```python
# 여러 열에 대해 각기 다른 집계 함수 적용
pt = pd.pivot_table(
    df,
    values=['수량', '금액'],
    index=['구분', '크기'],
    aggfunc={'수량': "mean", '금액': "mean"}
)
pt
```

```python
# 하나의 열에 대해 여러 집계 함수를 동시에 적용
pt = pd.pivot_table(
    df,
    values=['수량', '금액'],
    index=['구분', '크기'],
    aggfunc={'수량': "mean", '금액': ["min", "max", "mean"]}
)
pt
```
